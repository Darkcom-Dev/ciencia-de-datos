## Glosario de terminos.

### Inteligencia artificial:
Su enfoque es que la maquina entienda: Una imagen y sepa reconocerla, un sonido y sepa reconocerlo, un texto y sepa entenderlo

### Machine learning:
Su enfoque es que la maquina entienda: Una imagen y sepa reconocerla, un sonido y sepa reconocerlo, un texto y sepa entenderlo.

### Machine Learning:
Su enfoque es que la máquina sepa tomar decisiones por su cuenta, basado en lo que entiende y no en un algoritmo préprogramado previamente.
- Usa algoritmos algoritmos que hacen uso de datos ya procesados (modelos)
- Los algoritmos son regresión lineal, arboles de decisiones etc.

### Data Science:
Su enfoque es buscar conocimientos y perspectivas a través de los datos usando el método científico y técnicas estadísticas.
- Busca patrones, relaciones y tendencias en los datos
- Buscan la mayor precisión en los datos y son conscientes que no encontraran exactitud, por eso hablan en términos de tasas, promedios, etc.
- En resumen el Data scientist busca crear modelos de predicción y toma de decisiones.

### Data Analist:
Su enfoque es buscar conocimientos y hacer gráficos que permitan explicar fácilmente los datos
- Se encarga de la recolección, limpieza y análisis de los datos, buscando tendencias y perspectivas
- Usa métodos estadísticos para resumir y visualizar los datos.

- EDA: Exploratory Data Analisis

### Data Engineer:
Su enfoque es la preservación de los datos, estructuración y mantenibilidad de los datos, se encarga de la seguridad, guardado y procesamiento de datos.

### Exploración de Datos

Aqui entramos en el uso de librerias para la exploración de datos, Pandas, Numpy, Matplotlib, etc. Con el fin de depurar los datos y descubrir los nulos, vacios, inconsistencias, descubrir cuales es la tendencia, correlaciones etc.

### Limpieza

Despues de explorar los datos, es necesario aplicar limpieza a los datos, como descartar columnas innecesarias, posiblemente rellenar datos nan, n/a, null, con un promedio entre datos o definitivamente eliminar las filas, todo esto depende de la cantidad de datos comprometidos.

### Estandarizar los datos.

Hay datos redundantes como `Hombre, Mujer, M, F, Masculino, Femenino` estos datos deben ser consistentes y escoger un mismo par, tambien es necesario categorizar los datos.

### One hot Encoding.

Despues de la limpieza y la consistencia de los datos debes tomar todas categorias y partirlos en distintas columnas por categoria, esto se hace mediante **one hot encoding**.

### Particionado

Se toma un 70% aproximado de los datos para entrenamiento y el resto  para evaluación.

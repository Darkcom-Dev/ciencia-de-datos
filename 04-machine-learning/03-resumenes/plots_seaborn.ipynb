```python
import numpy as np

variacion = np.random.random([36,1])
variacion
```

```python
df_var = pd.DataFrame( variacion, columns=['variacion'])
df_var
```

```python
df_concatenado = pd.concat([df_l, df_var], axis=1)
df_concatenado
```

```python

df_grouped = df_concatenado.groupby(["mes y año",]).sum()["variacion"].reset_index()
df_grouped
```

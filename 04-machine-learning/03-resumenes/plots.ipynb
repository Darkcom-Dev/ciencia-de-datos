```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1)

Z = np.random.normal(0,1,100)
Z2 = np.random.normal(0,1,100)
X = np.arange(0,100)
diff = Z / Z2
ax1.plot(X, Z2, linestyle='dashed', color= 'green')
ax1.set_title('Visitors')
ax1.set_xlabel('Months')
ax1.set_ylabel('Visitors x 1k')
ax2.plot(X, Z, marker='o', color='orange')
ax2.set_title('Sales')
ax2.set_ylabel('Sales x 1k')
ax2.set_xlabel('Months')
ax3.plot(X, diff, color='blue')
ax3.set_ylabel('Difference')
```

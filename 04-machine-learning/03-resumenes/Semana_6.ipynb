# Regresion lineal

La regresion lineal busca una solucion analitica a datos tomados en la realidad, como en la realidad los datos no son exactos, generalmente forma una nube de puntos o dispersion en una grafica, pero por lo general la nube de puntos muestra una tendencia, generalmente una linea, sea logaritmica, exponencial o lineal.

Partiendo de que los datos tienden a formar una linea, la regresion lineal busca una ecuacion que genere la linea mas ajustada a la dispersion de los datos. la formula de la ecuacion de una linea es:


$y = mx + b$

- y = componente y de un punto
- m = pendiente
- x = componente x de un punto
- b = intersecto

Para hayar la pendiente es necesario hacer una resta entre 2 puntos:

$m = \frac{y_2 - y_1}{x_2 - x_1}$

Siendo el punto uno $(x_1,y_1)$ y el punto dos $(x_2, y_2)$

## Diferencia entre Funcion y Relacion

Hay representaciones en planos cartesianos que muestran lineas curvas como es el caso de las funciones exponenciales, logaritmicas y radicales. pero algunas funciones generan un reflejo como es el caso de las funciones exponenciales y las funciones que usan raiz cuadradas.

- Cuando el intersecto o linea de corte, toca dos puntos, estamos hablando de una relacion
- Cuando el intersecto toca solo un punto, estamos hablando de una funcion.

### Limites y deduccion de Derivada

Cuando se quiere hayar la pendiente, es necesario usar dos puntos como referencia, pero a veces se quiere que los puntos esten lo mas cercano posible, donde su distancia sea casi cero.

$lim x:h->0 = \frac{f(x+h) - f(x)}{x+h - x}$ esto es similar a: $m = \frac{y_2-y_1}{x_2-x_1}$

si $f(x) = x^2$

$lim h->0 = \frac{(x+h)^2 - x^2}{x + h -x}$

- luego $(x+h)^2$ se factoriza y quedaria asi: $\frac{x^2 + 2xh + h^2 - x^2}{x + h - x}$
- luego se cancelan los terminos y queda asi $\frac{h(2x + h)}{h}$
- luego se cancelan las h y queda $2x+h$
- finalmente como h tiende a cero se desprecia quedando solo $2x$
- 2x es el resultado de derivar $x^2$

$f(x) = X^2$
$f'(x) = 2X$



#### Derivadas.

La derivada siempre encuentra los puntos maximos y minimos, osea donde la pendiente es cero. En la deduccion anterior si graficamos veriamos exactamente eso:


````python
x = range(-100, 100)
y = [i ** 2 for i in x]
z = [i * 2 for i in x]

plt.plot(x, y)
plt.plot(x, z, color="red")
plt.show()
````

Se puede ver que la linea roja toca precisamente en el punto minimo, ahora vamos a graficar lo mismo con las siguientes funciones.

- $f(x) = \sin(x)$
- $f'(x) = \cos(x)$
- $f''(x) = -\sin(x)$
- $f'''(x) = -\cos(x)$

Si observa casi que cada vez que la linea negra toca un maximo o un minimo y su pendiente es cero, la linea roja está en cero en ese punto (perdon por la resolucion de trabajar con enteros). Lo mismo ocurre con la linea verde y azul, cuando la linea roja está en maximos o minimos y la pendiente es cero, la linea verde está en el punto cero, lo mismo ocurre con la linea azul que esta en cero cuando la linea verde está en máximos o minimos.

La linea roja esta en maximos o minimos cuando la pendiente es maxima sea subiendo o bajando, que por lo general es cuando la linea negra atraviesa el punto cero.

La función `fit()` de statsmodels.formula se utiliza para ajustar un modelo estadístico a los datos utilizando una fórmula especificada. La fórmula describe la **relación entre las variables dependientes e independientes** en el modelo.

Cuando se llama a la función `fit()`, se realiza el ajuste del modelo a los datos de acuerdo con la fórmula especificada. Esto implica estimar los **coeficientes** del modelo que mejor se ajustan a los datos observados. El resultado de la función `fit()` es un objeto Result que contiene los resultados del ajuste del modelo.

Aquí tienes un ejemplo de cómo se puede utilizar la función `fit()` de statsmodels.formula:


```python
import statsmodels.formula.api as smf

# Datos de ejemplo
data = {'x': [1, 2, 3, 4, 5], 'y': [2, 4, 6, 8, 10]}

# Crear una fórmula para el modelo
formula = 'y ~ x'

# Ajustar el modelo a los datos utilizando la fórmula
model = smf.ols(formula=formula, data=data).fit()

# Obtener los resultados del ajuste
results = model.summary()

# Mostrar los resultados
print(results)
```

En este ejemplo, se utiliza la función **ols()** de `statsmodels.formula.api` para crear un modelo de regresión lineal ordinaria. La fórmula `'y ~ x'` especifica que la variable dependiente **'y'** se relaciona linealmente con la variable independiente **'x'**. Luego, se utiliza la función **fit()** para ajustar el modelo a los datos utilizando la fórmula. Finalmente, se muestra un resumen de los resultados del ajuste utilizando el método **summary()** del objeto Result.

La función `fit()` es una parte fundamental de la librería statsmodels en Python y se utiliza para ajustar una **amplia gama de modelos estadísticos** a los datos.




#### Stats models formula o SMF

Esta libreria crea modelos a traves de una formula que pasas a traves de un string, a parte de OLS (minimos cuadrados ordinarios) que sirve para minimizar el error, existen otros modelos estadisitcos como:

- **Generalized Linear Models (GLM)**: Los modelos lineales generalizados son una extensión del modelo de regresión lineal que permite manejar diferentes distribuciones de errores y relaciones entre las variables. Algunos ejemplos de modelos GLM son el modelo de regresión logística, el modelo de regresión de Poisson y el modelo de regresión gamma.

- **Generalized Estimating Equations (GEE)**: Los modelos GEE son utilizados para análisis de datos longitudinales o correlacionados, donde se tiene en cuenta la estructura de correlación entre las observaciones. Estos modelos permiten ajustar datos con una varianza heterocedástica y una correlación de error.

- **Mixed Effects Models**: Los modelos de efectos mixtos, también conocidos como modelos de efectos aleatorios, son utilizados para analizar datos con estructura jerárquica o datos repetidos. Estos modelos tienen términos fijos y términos aleatorios, lo que permite modelar tanto los efectos fijos como las variaciones aleatorias en los datos.

- **Robust Linear Models**: Los modelos lineales robustos son utilizados cuando los supuestos de normalidad y homocedasticidad no se cumplen. Estos modelos utilizan métodos robustos para estimar los coeficientes del modelo y son menos sensibles a valores atípicos y violaciones de los supuestos.

Estos son solo algunos ejemplos de los modelos que se pueden ajustar utilizando la librería statsmodels.formula en Python. Cada uno de estos modelos tiene su propia sintaxis y opciones específicas al utilizar la función fit().

### The Box-Cox transformation

Logarithmic transformations are just one possible transformation to make our relationships more linear. Earlier, we mentioned powers (e.g. squares, cubes, square roots, etc.) as well as trigonometric functions. In some cases, choosing a transformation can be straightforward (e.g. the logarithm because it is easily interpretable); other times, it is much more difficult. A formal way to decide on which transformation to use is something called the [*Box-Cox*](https://www.isixsigma.com/tools-templates/normality/making-data-normal-using-box-cox-power-transformation/) criterion. This is a nifty mathematical function that outputs a number when you apply it to a variable. This number is usually represented by the Greek letter $\lambda$ (lambda). The interpretation of some common values of $\lambda$ is shown below:

| If you get a lambda value ($λ$) of... 	| then transform $Y$ using this function 	|
|-	|-	|
| -3 	| $Y^{-3} = \frac{1}{Y^3}$ 	|
| -2 	| $Y^{-2} = \frac{1}{Y^2}$	|
| -1 	| $Y^{-1} = \frac{1}{Y^1}$ 	|
| -0.5 	| $Y^{-0.5} = \frac{1}{Y^{0.5}} = \frac{1}{\sqrt{Y}}$ 	|
| 0 	| $log(Y)$ 	|
| 0.5 	| $Y^{0.5} = \sqrt{Y}$ 	|
| 1 	| $Y^1 = Y$ 	|
| 2 	| $Y^2$ 	|
| 3 	| $Y^3$ 	|


For instance, if you have a variable $Y$ and you got a $\lambda=2$ after you applied the Box-Cox function, then $Y^2$ should be approximately normally distributed. As a special case, when $\lambda=0$ we don't transform using $Y^0$ (which would give you the constant number 1), but rather using $\log(Y)$. In our case, we have a BC $\lambda=-0.23$, so it would seem that using the logarithmic transformation is sensible since the Box-Cox value is close to zero. Although it is very useful, the Box-Cox criterion is not perfect, so you should always double-check the QQ-plot after transforming your variable to confirm that it does seem to be normally distributed.

## Appendix: Elasticidades y semi-elasticidades

### Elasticidades

Rellamar nuestro primer modelo:

$$
\widehat{\log(price)}=6.7299 + 0.8368 \log(squareft{\_}living)
$$

Aseguremonos de usar la identidad $e^{\log(x)}=x$ Y el hecho es que $e^{(x+y)} = e^x e^y$ and $e^{px}=(e^x)^p$ para obtener

$$
\begin{aligned}
e^{\widehat{\log(price)}}&=e^{(6.7299 + 0.8368 \log({squareft{\_}living))}}\\
\widehat{price}&=e^{6.7299}e^{0.8368 \log({squareft{\_}living)}}\\
\widehat{price}&=e^{6.7299}(e^{\log({squareft{\_}living)}})^{0.8368}\\
\widehat{price}&=e^{6.7299}squareft{\_}living^{0.8368}
\end{aligned}
$$

Esto es una relacion no-lineal, entonces este no es tan directo como se dice "incrementando $squareft\_living$ por 1 significa que $\widehat{price}$ goes up by $x$". Sin embargo, Nosotros podemos intentar remarcar iste porcentaje en terminos de; i.e. how does a 1 percent increase in $squareft\_living$ affect $\widehat{price}$? Nosotros podemos ver esto al calcular el precio de la casa con solo un pie cuadrado:

$$
\begin{aligned}
\widehat{price}&=e^{6.7299}squareft{\_}living^{0.8368}\\
&=e^{6.7299}1^{0.8368}\\
&=837.063
\end{aligned}
$$

And comparing it with the price of a house with 1.01 square feet (thus, 1% more):

$$
\begin{aligned}
\widehat{price}&=e^{6.7299}squareft{\_}living^{0.8368}\\
&=e^{6.7299}1.01^{0.8368}\\
&=844.062
\end{aligned}
$$

The house with 1.01 square feet has a predicted price that is 1.008361198 times that of the other one, that is, this 1% increase in square footage is associated with a 0.8361% increase in predicted price. You can see that this percentage is very close to our regression model's $\beta_1=0.8368$. This is the reason why we can interpret $\beta_1$ as an elasticity in a log-log linear regression. In a slightly more general case, defining $S=1.01\times squareft{\_}living$, we have:

$$
\begin{aligned}
\widehat{price_S}&=e^{6.7299}S^{0.8368}\\
&=(e^{6.7299}squareft{\_}living^{0.8368})(1.01)^{0.8368}\\
&= (1.008361198) \widehat{price} \\
&\approx \beta_1 \widehat{price} \\
\end{aligned}
$$

which implies that

$$
\beta_1 \approx \frac{\widehat{price_S}}{\widehat{price}}
$$



### Semi-elasticities

Similar al la explicacion previa, tengamos el modelo

$$
\widehat{\log(price)}=\beta_0 + \beta_1 squareft{\_}living
$$

por lo tanto,

$$
\widehat{price}=e^{\beta_0} e^{\beta_1 squareft{\_}living}
$$

Si nosotros definimos $S=squareft{\_}living+1$ (nota: no un porcentaje de incremento, pero un incremento absoluto), entonces

$$\begin{aligned}
\widehat{price_S}&=e^{\beta_0} e^{\beta_1 S}\\
&=e^{\beta_0} e^{\beta_1 (squareft{\_}living+1)}\\
&=e^{\beta_0} (e^{(squareft{\_}living+1)})^{\beta_1}\\
&=e^{\beta_0} e^{\beta_1 squareft{\_}living}e^{\beta_1}\\
&=e^{\beta_1}\widehat{price}\\
\end{aligned}
$$

Desde cual este continua:

$$
\begin{aligned}
e^{\beta_1}&=\frac{\widehat{price_S}}{\widehat{price}}\\
\end{aligned}
$$

En nuestro segundo modelo (the log-level model), we obtained $\beta_1=0.0004$. Thus,

$$
\begin{aligned}
\frac{\widehat{price_S}}{\widehat{price}}&=e^{0.0004}\\
&=1.0004\\
&\approx \beta_1 + 1
\end{aligned}
$$

Y por lo tanto

$$
\beta_1 \approx \frac{\widehat{price_S}}{\widehat{price}} - 1
$$

El cual es semielastico (El porcentaje de cambio en la variable explicada que está asociada con un cambio absoluto de una unidad en la variable explanatoria).

# Kernels Vs Regresiones lineales alternos

Hay dos formas de obtener un modelo para crear predicciones, una de ellas ver como se comportan los datos en los gráficos de dispersion y elegir una regresion lineal adecuada.

La otra forma forzar que la nube de puntos se comporte como una linea usando un kernel y luego usar una regresion lineal simple.

## ¿Cual método es mas correcto?

Realmente no hay un método correcto y como consecuencia hay que probar distintas maneras para hacer la regresion, sin embargo, es mejor optar por la solucion que facilite el procedimiento y los calculos, por lo general es la solucion mas correcta.

## Complejidad Vs Simplicidad.

Al hacer la regresion lineal, no hay forma de saber si el modelo que hemos entrenado es el correcto. por eso hay que apegarnos a varias metricas para saber si los datos tienen relacion directa o indirecta.

Un modelo muy simple tiende a dar predicciones obvias que cualquier persona sin estudio puede predecir, esto deriva como resultado un $R^2$ muy bajo al rededor de 0.75

Por el contrario un modelo complejo tiene un $R^2$ muy alto, tendiendo al 0.95, pareciendo que es muy bueno prediciendo, pero que al comprobar resultados con datos nuevos, sus predicciones son muy regulares.

Lo ideal es que el modelo no sea ni muy simple ni muy complejo.

## Como identificar que a un modelo le falta complejidad.

Cuando separas los datos, en datos de entrenamiento y datos de prueba y comparas el $R^2$ de ambos conjuntos de datos, se puede obtener 3 escenarios:

1. Ideal: Cuando ambos el $R^2$ de los datos de entrenamiento y de prueba son muy similares.
2. Simple: Cuando el $R^2$ del entrenamiento es bajo y el $R^2$ de los test es alto, esto podria suponer que necesita aumentar la complejidad sumando una o dos columnas al modelo de regresion lineal multivariada.
3. Compleja: Cuando el $R^2$ del entrenamiento es alto y el $R^2$ de las pruebas es bajo. Esto quiere decir que nuestro modelo tiene una cantidad enorme de variables que hacen que el $R^2$ parezca ser muy alto, pero que no sirve, sugiriendo que hemos acomodado el modelo a lo que queremos y no a que sea cercano a la realidad. Para resolverlo se debe elegir bien las variables mas relevantes o elegir un modelo de regresion mas simple pero que aun asi se ajusta bien al modelo.

## Validacion cruzada o Cross Validation

El punto anterior nos puede dar una pista de si nuestro modelo es simple o complejo, pero como toman conjuntos de datos con la ley de Pareto del 80% y 20%, el 20% puede dar un $R^2$ diferente a los datos de entrenamiento.

Una mejor manera de validar los datos es separar los datos en 3 tercios, tomo los primeros 2 tercios y entreno, luego contrasto con el ultimo tercio y tomo ambos $R^2$, 

Luego repito el mismo proceso entrenando con los 2 ultimos tercios y contrastando con el primer tercio.

Por ultimo repito el mismo proceso entrenando el modelo con el primer y tercer tercio y contrastando con el segundo tercio.

Luego queda comparar el $R^2$ resultante de los 3 entrenamientos y ver si el resultado es parecido. Si es asi, entonces el modelo es consistente, si no entonces estamos eligiendo mal las variables o mal la regresion.

# Que es Fanning Out

En el contexto de la ciencia de datos, el término "fanning out" se refiere a un proceso en el que se amplía el conjunto de datos inicial para incluir más características o variables relevantes. Esto implica expandir o enriquecer los datos existentes para obtener una representación más completa y detallada de los mismos.

El fanning out en ciencia de datos puede tener varias aplicaciones, como:

- **Recopilación de datos adicionales**: Agregar nuevas fuentes de datos o variables a un conjunto de datos existente para obtener una imagen más completa o para incluir información adicional relevante.

- **Ingeniería de características**: Crear nuevas características o variables derivadas a partir de las características existentes, lo cual puede permitir un mejor rendimiento de los modelos de machine learning.

- **Extracción de características**: Utilizar técnicas de extracción de características para transformar datos no estructurados o de alta dimensionalidad en representaciones más compactas y significativas.

- **Generación de datos sintéticos**: Crear datos sintéticos para aumentar el tamaño o la diversidad de un conjunto de datos, lo cual puede ser útil en situaciones en las que el conjunto de datos existente es limitado o desequilibrado.

En resumen, el fanning out en ciencia de datos se refiere a la ampliación o enriquecimiento de un conjunto de datos existente mediante la adición de nuevas características, variables o información, con el objetivo de obtener una representación más completa y detallada de los datos.

# Pasty sintax

La sintaxis Pasty es una forma de escribir y formatear código en Python que se utiliza en algunos entornos de desarrollo y editores de texto. Pasty es una abreviatura de "Python Abstract Syntax Trees" (Árboles de Sintaxis Abstracta de Python) y se centra en la representación visual del código Python.

En la sintaxis Pasty, el código se organiza en bloques visuales utilizando sangrías y líneas en blanco. A diferencia de la sintaxis estándar de Python, que utiliza sangrías con espacios o tabulaciones, la sintaxis Pasty utiliza caracteres especiales como |, - y + para indicar la estructura del código.

Aquí hay un ejemplo simple de código Python escrito en sintaxis Pasty:

```pasty
def greet():
    |print("Hello,")
    |print("Welcome to Pasty syntax!")
    |
    +greet()
```

En este ejemplo, los bloques de código se indican mediante la línea vertical |, la línea horizontal - y el símbolo +. El bloque def greet(): indica una función llamada "greet" y los bloques de código indentados debajo de ella forman parte de la definición de la función. Los bloques de código print("Hello,") y print("Welcome to Pasty syntax!") están indentados debajo de la función greet() y se ejecutarán cuando se llame a la función.

La sintaxis Pasty se utiliza principalmente como una representación visual del código Python para ayudar a mejorar la legibilidad y la comprensión del código, especialmente en entornos de desarrollo específicos o editores de texto que admiten esta sintaxis.

$$y=w_0+w_1x_1+w_2x_2+...+w_nx_n$$

Primero necesitamos de una medida de qué tan bien (o mal) el modelo se ajusta a los datos de entrenamiento. Esta medida de evaluación (función de costo) es el error calculado entre la recta generada $\hat{y}$ (o el hiperplano) a los puntos reales. El entrenamiento del modelo será entonces encontrar los valores de $w_i$ que minimicen dicha función de costo. Entre las métricas más populares encontramos:

* Error medio absoluto (MAE)

$$MAE = \frac{1}{m}\sum_{i=1}^{m}|\hat{y}_i -y_i|$$

* Error cuadrático medio (MSE)

$$MSE=\frac{1}{m}\sum_{i=1}^{m}\left(\hat{y}_i -y_i\right)^2$$

* Raíz del error cuadrático medio (RMSE)

$$RMSE=\sqrt{\frac{1}{m}\sum_{i=1}^{m}\left(\hat{y}({\bf x})_i -y_i\right)^2}$$

Es hora de recurrir al tercer compartimento de la caja de herramientas para un algoritmo. Uno de los algoritmos más simples en el aprendizaje automático es el análisis de regresión, que se utiliza para determinar la fuerza de la relación entre variables. El análisis de regresión se presenta en muchas formas, incluidas la lineal, no lineal, logística y multilínea, pero primero veamos la regresión lineal.

La regresión lineal comprende una línea recta que divide tus puntos de datos en un diagrama de dispersión. El objetivo de la regresión lineal es dividir tus datos de manera que se minimice la distancia entre la línea de regresión y todos los puntos de datos en el diagrama de dispersión. Esto significa que si dibujaras una línea vertical desde la línea de regresión hasta cada punto de datos en el gráfico, la distancia agregada de cada punto equivaldría a la distancia más pequeña posible a la línea de regresión.



Figure 2: Linear regression line

La línea de regresión se traza en el diagrama de dispersión en la Figura 2. El término técnico para la línea de regresión es el hiperplano, y verás este término usado a lo largo de tu estudio de aprendizaje automático. Un hiperplano es prácticamente una línea de tendencia, y esto es precisamente cómo Google Sheets titula la regresión lineal en su menú de personalización de diagramas de dispersión.

Otra característica importante de la regresión es la pendiente, que se puede calcular convenientemente haciendo referencia al hiperplano. A medida que una variable aumenta, la otra variable aumentará en el valor promedio indicado por el hiperplano. Por lo tanto, la pendiente es muy útil para formular predicciones. Por ejemplo, si deseas estimar el valor de Bitcoin en 800 días, puedes ingresar 800 como tu coordenada x y consultar la pendiente para encontrar el valor y correspondiente representado en el hiperplano. #### Calculation Example

Aunque tu lenguaje de programación se encargará de esto automáticamente, es útil entender cómo se calcula la regresión lineal. Utilizaremos el siguiente conjunto de datos y fórmula para realizar la regresión lineal.

> El aporte que sigue lo ha hecho chatGPT
Dataset:
- $ X $ (días transcurridos): [100, 200, 300, 400, 500]
- $ Y $ (precio de Bitcoin): [800, 1200, 1600, 2000, 2400]

Fórmula de la regresión lineal:
$$ Y = mx + b $$

Donde:
- $ m $ es la pendiente de la línea de regresión.
- $ b $ es la intersección y.

Primero, calculamos la pendiente $ m $:

$$ m = \frac{n(\sum{XY}) - (\sum{X})(\sum{Y})}{n(\sum{X^2}) - (\sum{X})^2} $$

Donde:
- $ n $ es el número de puntos de datos (en este caso, 5).
- $ \sum{XY} $ es la suma de los productos de cada $ X $ con su correspondiente $ Y $.
- $ \sum{X} $ es la suma de todos los valores de $ X $.
- $ \sum{Y} $ es la suma de todos los valores de $ Y $.
- $ \sum{X^2} $ es la suma de los cuadrados de cada $ X $.

Sustituyendo los valores conocidos:
$$ n = 5 $$
$$ \sum{X} = 100 + 200 + 300 + 400 + 500 = 1500 $$
$$ \sum{Y} = 800 + 1200 + 1600 + 2000 + 2400 = 8000 $$
$$ \sum{XY} = (100 \times 800) + (200 \times 1200) + (300 \times 1600) + (400 \times 2000) + (500 \times 2400) = 3800000 $$
$$ \sum{X^2} = 100^2 + 200^2 + 300^2 + 400^2 + 500^2 = 225000 $$

Entonces, calculamos $ m $:
$$ m = \frac{(5 \times 3800000) - (1500 \times 8000)}{(5 \times 225000) - (1500)^2} $$
$$ m = \frac{(19000000) - (12000000)}{(1125000) - (2250000)} $$
$$ m = \frac{7000000}{-1125000} $$
$$ m = -6.22 $$

Ahora, calculamos $ b $ (la intersección y):

$$ b = \frac{\sum{Y} - m\sum{X}}{n} $$

Sustituyendo los valores conocidos:
$$ b = \frac{8000 - (-6.22 \times 1500)}{5} $$
$$ b = \frac{8000 + 9330}{5} $$
$$ b = \frac{17330}{5} $$
$$ b = 3466 $$

Por lo tanto, la ecuación de regresión lineal es:
$$ Y = -6.22X + 3466 $$

Esto significa que, según nuestra regresión lineal, el precio de Bitcoin disminuirá en aproximadamente $6.22 por día, comenzando en $3466 después de 100 días.

#### Logistic Regression

Una gran parte del análisis de datos se reduce a una pregunta simple: ¿es algo "A" o "B"? ¿Es "positivo" o "negativo"? ¿Esta persona es un "cliente potencial" o "no un cliente potencial"? El aprendizaje automático responde a tales preguntas a través de ecuaciones logísticas, y específicamente a través de lo que se conoce como la función sigmoide. La función sigmoide produce una curva en forma de "S" que puede convertir cualquier número y mapearlo en un valor numérico entre 0 y 1, pero lo hace sin llegar a esos límites exactos.

Una aplicación común de la función sigmoide se encuentra en la regresión logística.

La regresión logística adopta la función sigmoide para analizar datos y predecir clases discretas que existen en un conjunto de datos. Aunque la regresión logística comparte una similitud visual con la regresión lineal, técnicamente es una técnica de clasificación. Mientras que la regresión lineal aborda ecuaciones numéricas y forma predicciones numéricas para discernir relaciones entre variables, la regresión logística predice clases discretas.



Figure 6: An example of logistic regression

La regresión logística se utiliza típicamente para la clasificación binaria para predecir dos clases discretas, por ejemplo, embarazada o no embarazada. Para hacer esto, se utiliza la función sigmoide (mostrada a continuación) para calcular el resultado y convertir los resultados numéricos en una expresión de probabilidad entre 0 y 1.

$$y=\frac{1}{1+e^{-x}}$$

La función sigmoide logística se calcula como "1" dividido por "1" más "e" elevado a la potencia de menos "x", donde:
- $ x $ = el valor numérico que deseas transformar
- $ e $ = la constante de Euler, aproximadamente 2.718

En un caso binario, un valor de 0 representa ninguna posibilidad de ocurrencia, y 1 representa una posibilidad cierta de ocurrencia. El grado de probabilidad para los valores ubicados entre 0 y 1 se puede calcular según qué tan cerca estén de 0 (imposible) o 1 (posibilidad cierta) en el diagrama de dispersión.



Figure 7: A sigmoid function used to classify data points

Basándonos en las probabilidades encontradas, podemos asignar cada punto de datos a una de dos clases discretas. Como se ve en la Figura 7, podemos crear un punto de corte en 0.5 para clasificar los puntos de datos en clases. Los puntos de datos que registran un valor por encima de 0.5 se clasifican como Clase A, y cualquier punto de datos por debajo de 0.5 se clasifica como Clase B. Los puntos de datos que registran un resultado exacto de 0.5 son no clasificables, pero tales casos son raros debido al componente matemático de la función sigmoide.

También es importante tener en cuenta que esta fórmula por sí sola no produce el hiperplano que divide las categorías discretas como se ve en la Figura 6 anterior. La fórmula estadística para trazar el hiperplano logístico es algo más complicada y puede trazarse convenientemente utilizando tu lenguaje de programación.

Dada su fortaleza en la clasificación binaria, la regresión logística se utiliza en muchos campos, incluida la detección de fraudes, el diagnóstico de enfermedades, la detección de emergencias, la detección de incumplimientos de préstamos o para identificar correos electrónicos no deseados mediante la identificación de clases específicas, por ejemplo, no spam y spam. Sin embargo, la regresión logística también se puede aplicar a casos ordinales donde hay un número establecido de valores discretos, por ejemplo, soltero, casado y divorciado.

La regresión logística con más de dos valores de resultado se conoce como regresión logística multinomial, que se puede ver en la Figura 8.



Figure 8: An example of multinomial logistic regression

Dos consejos a recordar al realizar regresión logística son que los datos deben estar libres de valores faltantes y que todas las variables deben ser independientes entre sí. También debe haber datos suficientes para cada valor de resultado para garantizar una alta precisión. Un buen punto de partida serían aproximadamente 30-50 puntos de datos para cada resultado, es decir, 60-100 puntos de datos en total para la regresión logística binaria.

#### Support Vector Machine

Como una categoría avanzada de regresión, las máquinas de vectores de soporte (SVM, por sus siglas en inglés) se asemejan a la regresión logística pero con condiciones más estrictas. Con este fin, las SVM son superiores en la creación de líneas de límite de clasificación. Veamos cómo se ve esto en acción.



Figure 9: Logistic regression versus SVM

El gráfico de dispersión en la Figura 9 consiste en puntos de datos que son linealmente separables, y el hiperplano logístico (A) divide los puntos de datos en dos clases de manera que minimiza la distancia entre todos los puntos de datos y el hiperplano. La segunda línea, el hiperplano SVM (B), también separa los dos grupos, pero desde una posición de máxima distancia entre él y los dos grupos.

También notarás un área gris que denota el margen, que es la distancia entre el hiperplano y el punto de datos más cercano, multiplicado por dos. El margen es una característica clave de las SVM y es importante porque ofrece soporte adicional para manejar nuevos puntos de datos que puedan infringir en un hiperplano de regresión logística. Para ilustrar este escenario, consideremos el mismo gráfico de dispersión con la inclusión de un nuevo punto de datos.



Figure 10: A new data point is added to the scatterplot

El nuevo punto de datos es un círculo, pero está ubicado incorrectamente en el lado izquierdo del hiperplano de regresión logística (designado para estrellas). Sin embargo, el nuevo punto de datos sigue estando correctamente ubicado en el lado derecho del hiperplano SVM (designado para círculos) gracias al amplio "soporte" proporcionado por el margen.



Figure 11: Mitigating anomalies

Otro caso de aplicación útil de SVM es para mitigar anomalías. Una limitación de la regresión logística estándar es que se esfuerza por ajustar anomalías (como se ve en el diagrama de dispersión con la estrella en la esquina inferior derecha en la Figura 11). Sin embargo, SVM es menos sensible a tales puntos de datos y en realidad minimiza su impacto en la ubicación final de la línea de límite. En la Figura 11, podemos ver que la Línea B (hiperplano SVM) es menos sensible a la estrella anómala en el lado derecho. Por lo tanto, SVM puede utilizarse como un método para combatir anomalías.

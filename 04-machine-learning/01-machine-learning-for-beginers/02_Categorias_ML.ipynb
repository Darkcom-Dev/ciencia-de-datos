# ML CATEGORIAS

El aprendizaje automático incorpora varios cientos de algoritmos basados en estadísticas y elegir el algoritmo adecuado o la combinación correcta de algoritmos para el trabajo es un desafío constante para cualquier persona que trabaje en este campo. Pero antes de examinar algoritmos específicos, es importante comprender las tres categorías principales del aprendizaje automático. Estas tres categorías son supervisado, no supervisado y de refuerzo.

#### Aprendizaje supervisado

Como la primera rama del aprendizaje automático, el aprendizaje supervisado se concentra en aprender patrones mediante la conexión de la relación entre variables y resultados conocidos, trabajando con conjuntos de datos etiquetados. El aprendizaje supervisado funciona alimentando a la máquina datos de muestra con diversas características (representadas como "X") y el valor correcto de salida de los datos (representado como "y"). El hecho de que los valores de salida y características sean conocidos califica al conjunto de datos como "etiquetado". El algoritmo luego descifra los patrones que existen en los datos y crea un modelo que puede reproducir las mismas reglas subyacentes con nuevos datos.

Por ejemplo, para predecir la tasa de mercado para la compra de un automóvil usado, un algoritmo supervisado puede formular predicciones analizando la relación entre atributos del automóvil (incluido el año de fabricación, la marca del automóvil, el kilometraje, etc.) y el precio de venta de otros automóviles vendidos según datos históricos. Dado que el algoritmo supervisado conoce el precio final de otros automóviles vendidos, puede trabajar hacia atrás para determinar la relación entre las características del automóvil y su valor.

Una vez que la máquina descifra las reglas y los patrones de los datos, crea lo que se conoce como un modelo: una ecuación algorítmica para producir un resultado con nuevos datos basados en las reglas derivadas de los datos de entrenamiento. Una vez que el modelo está preparado, se puede aplicar a nuevos datos y probar su precisión. Después de que el modelo haya superado las etapas de datos de entrenamiento y de prueba, está listo para ser aplicado y utilizado en el mundo real.

En el Capítulo 13, crearemos un modelo para predecir los valores de las casas donde y es el precio real de la casa y X son las variables que impactan en y, como el tamaño del terreno, la ubicación y el número de habitaciones. A través del aprendizaje supervisado, crearemos una regla para predecir y (valor de la casa) basada en los valores dados de varias variables (X).

Ejemplos de algoritmos de aprendizaje supervisado incluyen análisis de regresión, árboles de decisión, k-vecinos más cercanos, redes neuronales y máquinas de vectores de soporte. Cada una de estas técnicas será presentada más adelante en el libro.

#### Aprendizaje no supervisado

En el caso del aprendizaje no supervisado, no todas las variables y patrones de datos están clasificados. En su lugar, la máquina debe descubrir patrones ocultos y crear etiquetas a través del uso de algoritmos de aprendizaje no supervisado. El algoritmo de agrupamiento k-means es un ejemplo popular de aprendizaje no supervisado. Este algoritmo simple agrupa los puntos de datos que se encuentran a poseer características similares como se muestra en la Figura 1.

Figure 1: Example of k-means clustering, a popular unsupervised learning technique

El aprendizaje no supervisado implica descubrir patrones en los datos sin clasificar variables ni patrones de datos. Por ejemplo, al agrupar los puntos de datos basados en el comportamiento de compra de las PYME y las grandes empresas, es probable que aparezcan dos grupos. Esto se debe a que las PYME y las grandes empresas tienden a tener hábitos de compra diferentes. Por ejemplo, al comprar infraestructura en la nube, los recursos básicos de alojamiento en la nube y una Red de Entrega de Contenidos (CDN) pueden ser suficientes para la mayoría de las PYME.

Los algoritmos de aprendizaje no supervisado, como el agrupamiento k-means, pueden identificar estos dos grupos de clientes sin etiquetas específicas que clasifiquen a la empresa como pequeña, mediana o grande.

La ventaja del aprendizaje no supervisado es que permite descubrir patrones en los datos que desconocíamos que existían, como la presencia de dos tipos principales de clientes. Además, las técnicas de agrupamiento, como el agrupamiento k-means, pueden proporcionar la base para realizar análisis adicionales después de descubrir los grupos discretos.

En la industria, el aprendizaje no supervisado es especialmente poderoso en la detección de fraudes, donde los ataques más peligrosos suelen ser aquellos que aún no están clasificados. Un ejemplo real es DataVisor, que básicamente construyó su modelo de negocio basado en el aprendizaje no supervisado.

DataVisor protege a los clientes de actividades fraudulentas en línea, como spam, reseñas falsas, instalaciones de aplicaciones falsas y transacciones fraudulentas. Mientras que los servicios tradicionales de protección contra fraudes utilizan modelos de aprendizaje supervisado y motores de reglas, DataVisor utiliza el aprendizaje no supervisado, lo que les permite detectar categorías no clasificadas de ataques en sus primeras etapas.

Por ejemplo, pueden identificar células durmientes (que imitan usuarios legítimos) a través de su asociación con cuentas maliciosas. Los algoritmos de agrupamiento, como el k-means, pueden generar estos grupos sin un conjunto de datos de entrenamiento completo.

El aprendizaje no supervisado se abordará más adelante en este libro, específicamente en análisis de agrupamiento. Otros ejemplos de aprendizaje no supervisado incluyen el análisis de asociación, el análisis de redes sociales y algoritmos de reducción de dimensionalidad.

#### Aprendizaje por Refuerzo

El aprendizaje por refuerzo es la tercera y más avanzada categoría de algoritmos en el aprendizaje automático. A diferencia del aprendizaje supervisado y no supervisado, el aprendizaje por refuerzo mejora continuamente su modelo aprovechando la retroalimentación de iteraciones anteriores. Esto difiere del aprendizaje supervisado y no supervisado, que alcanzan un punto final indefinido después de que se formula un modelo a partir de los segmentos de datos de entrenamiento y prueba.

El aprendizaje por refuerzo puede ser complicado y probablemente se explique mejor mediante una analogía con un videojuego. A medida que un jugador avanza a través del espacio virtual de un juego, aprende el valor de varias acciones bajo diferentes condiciones y se familiariza más con el campo de juego. Esas acciones aprendidas luego informan e influencian el comportamiento posterior del jugador y su desempeño mejora inmediatamente en función de su aprendizaje y experiencia pasada.

El aprendizaje por refuerzo es muy similar, donde los algoritmos se configuran para entrenar el modelo a través de un aprendizaje continuo. Un modelo de aprendizaje por refuerzo estándar tiene criterios de rendimiento medibles donde las salidas no están etiquetadas, en su lugar, se les asigna una calificación. En el caso de los vehículos autónomos, evitar un accidente asignará una puntuación positiva y en el caso del ajedrez, evitar la derrota también recibirá una puntuación positiva.

Un ejemplo algorítmico específico de aprendizaje por refuerzo es el Q-learning. En el Q-learning, se comienza con un entorno de estados establecido, representado por el símbolo "S". En el juego de Pac-Man, los estados podrían ser los desafíos, obstáculos o caminos que existen en el juego. Puede haber una pared a la izquierda, un fantasma a la derecha y una píldora de poder arriba, cada uno representando diferentes estados.

El conjunto de acciones posibles para responder a estos estados se denomina "A". En el caso de Pac-Man, las acciones están limitadas a movimientos hacia la izquierda, derecha, arriba y abajo, así como múltiples combinaciones de estas.

El tercer símbolo importante es "Q". Q es el valor inicial y tiene un valor inicial de "0".
A medida que Pac-Man explora el espacio dentro del juego, dos cosas principales sucederán:
1) Q disminuye cuando ocurren cosas negativas después de un estado/acción dado
2) Q aumenta cuando ocurren cosas positivas después de un estado/acción dado
En el Q-learning, la máquina aprenderá a emparejar la acción para un estado dado que genera o mantiene el nivel más alto de Q. Aprenderá inicialmente a través del proceso de movimientos (acciones) aleatorios bajo diferentes condiciones (estados).

La máquina registrará sus resultados (recompensas y penalizaciones) y cómo impactan su nivel de Q y almacenará esos valores para informar y optimizar sus acciones futuras.

Aunque esto parece lo suficientemente simple, la implementación es una tarea mucho más difícil y está más allá del alcance de una introducción absoluta al aprendizaje automático.

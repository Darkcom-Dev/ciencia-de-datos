# ¿Qué es Machine Learning?

En 1959, IBM publicó un artículo en el Jornal de Investigación y Desarrollo de IBM, escrito por Arthur Samuel, que exploraba el uso del aprendizaje de máquina en juegos de damas. Aunque no fue la primera publicación en usar el término "machine learning", Samuel es reconocido por acuñar y definir el concepto de aprendizaje de máquina como lo conocemos hoy en día. Su artículo, "Algunos estudios en aprendizaje de máquina usando el juego de damas", marcó un hito en la historia al abordar la posibilidad de que las computadoras pudieran aprender a jugar mejor al juego que las personas que programaban el software. Este trabajo también señaló el comienzo de los esfuerzos humanos por impartir habilidades de aprendizaje a las máquinas creadas por el hombre.

Arthur Samuel introdujo el término "Machine Learning" en su artículo como un subcampo de las ciencias de la computación que permite aprender sin ser programado explícitamente. Esta definición sigue siendo ampliamente aceptada casi seis décadas después. Aunque no está directamente mencionado en la definición de Samuel, una característica clave del aprendizaje de máquina es el concepto de autoaprendizaje, que se refiere a la aplicación de modelos estadísticos para detectar patrones y mejorar.

Arthur Samuel describió el aprendizaje de máquina como la capacidad de aprender sin ser explícitamente programado, basado en datos y evidencia empírica. Aunque no implica que las máquinas formulen decisiones sin programación previa, el aprendizaje de máquina depende en gran medida de la programación de computadoras. Samuel señaló que las máquinas no necesitan comandos de entrada directos para realizar una tarea, sino datos de entrada.

El proceso de entrada de comandos es directo y genera una respuesta inmediata. Por ejemplo, al ingresar `2+2` en Python, se obtiene el resultado `4`.

En contraste, los datos de entrada se envían a la máquina, se selecciona un algoritmo, se configuran los hiperparámetros y se indica a la máquina que realice su análisis. La máquina identifica patrones en los datos mediante ensayo y error, y forma un modelo de datos que se utiliza para predecir valores futuros.

Aunque hay una relación entre el programador y la máquina, operan en un nivel diferente en comparación con la programación tradicional. Esto se debe a que las decisiones se basan en la experiencia y se asemejan al proceso de toma de decisiones humanas.

Por ejemplo, después de analizar los hábitos de visualización en YouTube de los científicos de datos, la máquina identifica una fuerte relación entre ellos y los videos de gatos. Posteriormente, identifica patrones entre las características físicas de los jugadores de béisbol y su probabilidad de ganar el premio al Jugador Más Valioso (MVP). En ambos casos, la máquina no fue programada explícitamente para producir un resultado directo, sino que se alimentó con datos de entrada y se configuraron algoritmos.

Puede compararse la construcción de un modelo de datos con el entrenamiento de un perro guía. A través de un entrenamiento especializado, los perros guía aprenden a responder en diversas situaciones. De manera similar, los modelos de aprendizaje automático se entrenan para tomar decisiones basadas en experiencias pasadas.

Por ejemplo, se puede crear un modelo que detecte mensajes de correo electrónico no deseado. Aunque en esta etapa no se realiza aprendizaje automático, se pueden identificar dos procesos: comando > acción y datos > modelo > acción. Para incorporar el aprendizaje automático, se cambia "comando" por "datos" y se agrega "modelo" para producir una acción (salida). Los datos comprenden correos electrónicos de muestra y el modelo consta de reglas basadas en estadísticas. Después de entrenar y probar el modelo, existe la posibilidad de que se produzcan predicciones inexactas. Sin embargo, el aprendizaje automático puede adaptar y modificar suposiciones mediante su proceso de tres pasos, reaccionando ante los errores.

### Training & Test Data

En el aprendizaje automático, los datos se dividen en datos de entrenamiento y datos de prueba. Los datos de entrenamiento se utilizan para desarrollar el modelo inicial y ajustar las reglas, como excluir correos electrónicos legítimos que se identificaron erróneamente como spam. Una vez que el modelo se ajusta y se prueba con éxito, se aplica a los datos de prueba para confirmar su precisión.

La diferencia entre el aprendizaje automático y la programación tradicional se vuelve evidente a medida que se entiende el proceso de autoaprendizaje y su aplicación en situaciones más complejas.

Además, es crucial comprender cómo el aprendizaje automático se relaciona con otras disciplinas, como la ciencia de datos y la minería de datos. Mientras que la informática abarca el diseño y uso de computadoras, la ciencia de datos se centra en extraer conocimientos de los datos mediante sistemas computacionales. El aprendizaje automático es una parte integral de la ciencia de datos y se diferencia de la programación tradicional al utilizar datos para ajustar modelos en lugar de instrucciones explícitas.

Sobresaliendo de la informática y la ciencia de datos como la tercera muñeca matrioska está la inteligencia artificial. La inteligencia artificial, o IA, abarca la capacidad de las máquinas para realizar tareas inteligentes y cognitivas. Comparable a como la Revolución Industrial dio lugar a una era de máquinas que podían simular tareas físicas, la IA está impulsando el desarrollo de máquinas capaces de simular habilidades cognitivas.

Aunque aún es amplio pero dramáticamente más especializado que la informática y la ciencia de datos, la IA contiene numerosos subcampos que son populares hoy en día. Estos subcampos incluyen búsqueda y planificación, razonamiento y representación del conocimiento, percepción, procesamiento del lenguaje natural (NLP) y, por supuesto, el aprendizaje automático. El aprendizaje automático se extiende a otros campos de la IA, incluido el procesamiento del lenguaje natural y la percepción, mediante el uso compartido de algoritmos de autoaprendizaje.

Para estudiantes interesados en IA, el aprendizaje automático ofrece un excelente punto de partida, ya que ofrece un enfoque más estrecho y práctico en comparación con la ambigüedad conceptual de la IA. Los algoritmos del aprendizaje automático también se pueden aplicar en otras disciplinas, incluida la percepción y el procesamiento del lenguaje natural. Además, un título de maestría es adecuado para desarrollar cierto nivel de experiencia en aprendizaje automático, pero es posible que necesites un doctorado para avanzar verdaderamente en IA.

El aprendizaje automático también se superpone con la minería de datos, una disciplina hermana que se centra en descubrir patrones en grandes conjuntos de datos. Algoritmos populares, como el agrupamiento k-means, el análisis de asociación y el análisis de regresión, se aplican tanto en la minería de datos como en el aprendizaje automático para analizar datos. Sin embargo, mientras que el aprendizaje automático se enfoca en el proceso incremental de autoaprendizaje y modelado de datos para formar predicciones sobre el futuro, la minería de datos se centra en limpiar grandes conjuntos de datos para extraer información valiosa del pasado.

La diferencia entre la minería de datos y el aprendizaje automático se puede explicar mediante una analogía de dos equipos de arqueólogos. El primer equipo está formado por arqueólogos que se enfocan en eliminar escombros que obstruyen los objetos valiosos, ocultándolos de la vista directa. Sus objetivos principales son excavar el área, encontrar nuevos descubrimientos valiosos y luego empacar su equipo y mudarse. Al día siguiente, volarán a otro destino exótico para comenzar un nuevo proyecto sin relación con el sitio que excavaron el día anterior.

El segundo equipo también excava sitios históricos, pero estos arqueólogos utilizan una metodología diferente. Se abstienen deliberadamente de excavar el pozo principal durante varias semanas. Durante ese tiempo, visitan otros sitios arqueológicos relevantes en el área y examinan cómo se excavó cada sitio. Después de regresar al sitio de su propio proyecto, aplican este conocimiento para excavar pozos más pequeños que rodean el pozo principal.

Los arqueólogos luego analizan los resultados. Después de reflexionar sobre su experiencia excavando un pozo, optimizan sus esfuerzos para excavar el siguiente. Esto incluye predecir el tiempo que tomará excavar un pozo, comprender la varianza y los patrones encontrados en el terreno local y desarrollar nuevas estrategias para reducir errores y mejorar la precisión de su trabajo. A partir de esta experiencia, pueden optimizar su enfoque para formar un modelo estratégico para excavar el pozo principal.

Si aún no está claro, el primer equipo se adhiere a la minería de datos y el segundo equipo al aprendizaje automático. A nivel micro, tanto la minería de datos como el aprendizaje automático parecen similares y utilizan muchas de las mismas herramientas.

Ambos equipos ganan la vida excavando sitios históricos para descubrir objetos valiosos. Pero en la práctica, su metodología es diferente. El equipo de aprendizaje automático se enfoca en dividir su conjunto de datos en datos de entrenamiento y datos de prueba para crear un modelo y mejorar las predicciones futuras basadas en la experiencia anterior. Mientras tanto, el equipo de minería de datos se concentra en excavar el área objetivo de la manera más efectiva posible, sin usar un modelo de autoaprendizaje, antes de pasar al siguiente trabajo de limpieza.

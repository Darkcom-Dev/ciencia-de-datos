# THE ML TOOLBOX

Una forma útil de aprender un nuevo área temática es mapear y visualizar los materiales y herramientas esenciales dentro de una caja de herramientas.

Si estuvieras armando una caja de herramientas para construir sitios web, por ejemplo, primero incluirías una selección de lenguajes de programación. Esto incluiría lenguajes frontend como HTML, CSS y JavaScript, uno o dos lenguajes de programación backend basados en preferencias personales, y por supuesto, un editor de texto. Podrías incluir también un constructor de sitios web como WordPress y luego tener otro compartimento lleno de alojamiento web, DNS y tal vez algunos nombres de dominio que has comprado recientemente.

Esta lista no es exhaustiva, pero a partir de esta lista general, puedes comenzar a comprender mejor qué herramientas necesitas dominar para convertirte en un desarrollador web exitoso.

Ahora vamos a desempaquetar la caja de herramientas para el aprendizaje automático.

#### Compartment 1: Data

En el primer compartimento se encuentra tu datos. Los datos constituyen las variables de entrada necesarias para formar una predicción. Los datos vienen en muchas formas, incluyendo datos estructurados y no estructurados. Como principiante, se recomienda que comiences con datos estructurados. Esto significa que los datos están definidos y etiquetados (con un esquema) en una tabla, como se muestra aquí:

|Date|Bitcoin Price|No. of Days Transpired|
|----|-------------|----------------------|
|19-05-2015|234.31|1|
|14-01-2016|431.76|240|
|09-07-2016|652.14|417|
|15-01-2017|817.26|607|
|24-05-2017|2358.96|736|

Antes de proceder, primero quiero explicar la anatomía de un conjunto de datos tabular. Un conjunto de datos tabular (basado en tablas) contiene datos organizados en filas y columnas. En cada columna se encuentra una característica. Una característica también se conoce como una variable, una dimensión o un atributo, pero todos significan lo mismo.

Cada fila individual representa una sola observación de una característica/variable dada. A veces, las filas se denominan casos o valores, pero en este libro utilizaremos el término "fila".

Dentro del primer compartimento de la caja de herramientas se encuentran una variedad de gráficos de dispersión, que incluyen gráficos de 2-D, 3-D y 4-D. Un gráfico de dispersión 2-D consta de un eje vertical (conocido como el eje y) y un eje horizontal (conocido como el eje x) y proporciona el lienzo gráfico para trazar una serie de puntos, conocidos como puntos de datos.

Cada punto de datos en el gráfico de dispersión representa una observación del conjunto de datos, con los valores de X trazados en el eje x y los valores de y trazados en el eje y.

#### Compartment 2: Infrastructure

El segundo compartimento de la caja de herramientas contiene tu infraestructura, que consiste en plataformas y herramientas para procesar datos. Como principiante en aprendizaje automático, es probable que estés utilizando una aplicación web (como Jupyter Notebook) y un lenguaje de programación como Python. Luego están una serie de bibliotecas de aprendizaje automático, incluyendo NumPy, Pandas y Scikit-learn, que son compatibles con Python. Las bibliotecas de aprendizaje automático son una colección de rutinas de programación precompiladas utilizadas con frecuencia en el aprendizaje automático.

También necesitarás una máquina desde la cual trabajar, en forma de una computadora o un servidor virtual. Además, es posible que necesites bibliotecas especializadas para visualización de datos como Seaborn y Matplotlib, o un programa de software independiente como Tableau, que admite una variedad de técnicas de visualización, incluidos gráficos, mapas y otras opciones visuales.

Con tu infraestructura desplegada sobre la mesa (hipotéticamente, por supuesto), ahora estás listo para ponerte a trabajar en la construcción de tu primer modelo de aprendizaje automático. El primer paso es encender tu computadora. Tanto las computadoras portátiles como las de escritorio son adecuadas para trabajar con conjuntos de datos más pequeños. Luego, deberás instalar un entorno de programación, como Jupyter Notebook, y un lenguaje de programación, que para la mayoría de los principiantes es Python.

Python es el lenguaje de programación más ampliamente utilizado para el aprendizaje automático porque:
a) Es fácil de aprender y operar,
b) Es compatible con una variedad de bibliotecas de aprendizaje automático, y
c) Puede usarse para tareas relacionadas, incluida la recolección de datos (web scraping) y el procesamiento de datos (Hadoop y Spark).

Otros lenguajes populares para el aprendizaje automático incluyen C y C++. Si tienes habilidades con C y C++, tiene sentido continuar con lo que ya conoces. C y C++ son los lenguajes de programación predeterminados para el aprendizaje automático avanzado porque pueden ejecutarse directamente en una GPU (Unidad de Procesamiento Gráfico). Python necesita ser convertido primero antes de poder ejecutarse en una GPU, pero llegaremos a esto y a qué es una GPU más adelante en el capítulo.

A continuación, los usuarios de Python suelen instalar las siguientes bibliotecas: NumPy, Pandas y Scikit-learn. NumPy es una biblioteca gratuita y de código abierto que te permite cargar y trabajar eficientemente con conjuntos de datos grandes, incluyendo la gestión de matrices.

Scikit-learn proporciona acceso a una variedad de algoritmos populares, incluida la regresión lineal, el clasificador de Bayes y las máquinas de vectores de soporte.

Finalmente, Pandas permite que tus datos se representen en una hoja de cálculo virtual que puedes controlar mediante código. Comparte muchas de las mismas características que Microsoft Excel en el sentido de que te permite editar datos y realizar cálculos. De hecho, el nombre Pandas proviene del término "datos de panel", que se refiere a su capacidad para crear una serie de paneles, similares a las "hojas" en Excel. Pandas también es ideal para importar y extraer datos de archivos CSV.




Figure 4: Previewing a table in Jupyter Notebook using Pandas

En resumen, los usuarios pueden aprovechar estas tres bibliotecas para:

1) Cargar y trabajar con un conjunto de datos a través de NumPy.
2) Limpiar y realizar cálculos en los datos, y extraer datos de archivos CSV con Pandas.
3) Implementar algoritmos con Scikit-learn.

Para los estudiantes que buscan opciones de programación alternativas (más allá de Python, C y C++), otros lenguajes de programación relevantes para el aprendizaje automático incluyen R, MATLAB y Octave.

R es un lenguaje de programación gratuito y de código abierto optimizado para operaciones matemáticas, y propicio para construir matrices y funciones estadísticas, que están integradas directamente en las bibliotecas de R. Aunque R se usa comúnmente para análisis de datos y minería de datos, también admite operaciones de aprendizaje automático.

MATLAB y Octave son competidores directos de R. MATLAB es un lenguaje de programación comercial y propietario. Es fuerte en la resolución de ecuaciones algebraicas y también es un lenguaje de programación rápido de aprender.

MATLAB se utiliza ampliamente en ingeniería eléctrica, ingeniería química, ingeniería civil e ingeniería aeronáutica. Sin embargo, los científicos de la computación e ingenieros en computación tienden a no depender tanto de MATLAB, especialmente en tiempos recientes. En aprendizaje automático, MATLAB se usa más a menudo en la academia que en la industria. Por lo tanto, aunque puedes ver MATLAB en cursos en línea, especialmente en Coursera, esto no significa que se use comúnmente en la práctica. Sin embargo, si vienes de un fondo en ingeniería, MATLAB es ciertamente una opción lógica.

Por último, Octave es esencialmente una versión gratuita de MATLAB desarrollada en respuesta a MATLAB por la comunidad de código abierto.

#### Advanced Toolbox

Hasta ahora hemos examinado la caja de herramientas para un principiante típico, pero ¿qué pasa con un usuario avanzado? ¿Cómo se vería su caja de herramientas? Si bien puede tomar un tiempo antes de que trabajes con el conjunto de herramientas avanzadas, no está de más echar un vistazo rápido.

La caja de herramientas para un aprendiz avanzado se asemeja a la del principiante, pero naturalmente viene con un espectro más amplio de herramientas y, por supuesto, datos. Una de las mayores diferencias entre un principiante y un aprendiz avanzado es el tamaño de los datos que manejan y operan. Los principiantes comienzan naturalmente trabajando con conjuntos de datos pequeños que son fáciles de manejar y que se pueden descargar directamente en el escritorio como un simple archivo CSV. Sin embargo, los aprendices avanzados estarán ansiosos por abordar conjuntos de datos masivos, cercanos a los del big data.

#### Compartment 1: Big Data

El término "big data" se utiliza para describir un conjunto de datos que, debido a su valor, variedad, volumen y velocidad, desafía los métodos convencionales de procesamiento y sería imposible de procesar para un ser humano sin la ayuda de una máquina avanzada. El big data no tiene una definición exacta en términos de tamaño o número total de filas y columnas. En este momento, los petabytes califican como big data, pero los conjuntos de datos están creciendo cada vez más grandes a medida que encontramos nuevas formas de recopilar y almacenar datos de manera eficiente y económica. Y con el big data también vienen un mayor ruido y estructuras de datos complicadas. Por lo tanto, una gran parte de trabajar con big data consiste en depurar: el proceso de refinar tu conjunto de datos antes de construir tu modelo, lo cual se cubrirá en el próximo capítulo.

#### Compartimiento 2: Infraestructura

Después de depurar el conjunto de datos, el siguiente paso es sacar tu equipo de aprendizaje automático. En términos de herramientas, no hay sorpresas. Los estudiantes avanzados siguen utilizando las mismas bibliotecas de aprendizaje automático, lenguajes de programación y entornos de programación que los principiantes.

Sin embargo, dado que los estudiantes avanzados ahora están lidiando con hasta petabytes de datos, se requiere una infraestructura robusta. En lugar de depender de la CPU de una computadora personal, los estudiantes avanzados suelen recurrir a la computación distribuida y a un proveedor de servicios en la nube como Amazon Web Services (AWS) para ejecutar su procesamiento de datos en lo que se conoce como una Instancia de Procesamiento Gráfico (GPU).

Las tarjetas GPU se agregaron originalmente a las placas base de PC y a consolas de videojuegos como PlayStation 2 y Xbox con fines de juego. Se desarrollaron para acelerar la creación de imágenes con millones de píxeles cuyos fotogramas necesitaban ser constantemente recalculados para mostrar la salida en menos de un segundo. Para 2005, las tarjetas GPU se producían en cantidades tan grandes que su precio había caído drásticamente y esencialmente se habían convertido en una mercancía. Aunque son muy populares en la industria de los videojuegos, la aplicación de tales chips informáticos en el ámbito del aprendizaje automático no se entendió ni se realizó completamente hasta hace poco.

En su novela de 2016, "The Inevitable: Understanding the 12 Technological Forces That Will Shape Our Future", Kevin Kelly, editor ejecutivo fundador de la revista Wired, explica que en 2009, Andrew Ng y un equipo de la Universidad de Stanford descubrieron cómo vincular clústeres GPU económicos para ejecutar redes neuronales que consisten en cientos de millones de conexiones de nodos.

"Los procesadores tradicionales requerían varias semanas para calcular todas las posibilidades en cascada en una red neuronal con cien millones de parámetros. Ng encontró que un clúster de GPUs podía lograr lo mismo en un día". Como un chip de cómputo paralelo especializado, las instancias de GPU pueden realizar muchas más operaciones de punto flotante por segundo que una CPU, lo que permite soluciones mucho más rápidas con álgebra lineal y estadísticas que con una CPU.

Es importante tener en cuenta que C y C++ son los lenguajes preferidos para editar y realizar operaciones matemáticas directamente en la GPU. Sin embargo, también se puede utilizar Python y convertirlo a C en combinación con TensorFlow de Google.

Aunque es posible ejecutar TensorFlow en la CPU, se puede obtener hasta unas 1000 veces más de rendimiento utilizando la GPU. Desafortunadamente para los usuarios de Mac, TensorFlow solo es compatible con la tarjeta GPU Nvidia, que ya no está disponible con Mac OS X. Los usuarios de Mac aún pueden ejecutar TensorFlow en su CPU, pero necesitarán diseñar un parche/controlador externo o ejecutar su carga de trabajo en la nube para acceder a la GPU. Amazon Web Services, Microsoft Azure, Alibaba Cloud, Google Cloud Platform y otros proveedores de servicios en la nube ofrecen recursos de GPU de pago por uso, que pueden comenzar de forma gratuita a través de un programa de prueba gratuita.

Actualmente, Google Cloud Platform se considera una opción líder para recursos de GPU en términos de rendimiento y precio. En 2016, Google también anunció que lanzaría públicamente una Unidad de Procesamiento Tensorial diseñada específicamente para ejecutar TensorFlow, que ya se utiliza internamente en Google.

#### Compartimiento 3: Algoritmos Avanzados

Para completar este capítulo, echemos un vistazo al tercer compartimiento de la caja de herramientas avanzada que contiene algoritmos de aprendizaje automático.

Para analizar conjuntos de datos grandes, los estudiantes avanzados trabajan con una gran cantidad de algoritmos avanzados que incluyen modelos de Markov, máquinas de vectores de soporte y Q-learning, así como una serie de algoritmos simples como los que se encuentran en la caja de herramientas para principiantes. Pero la familia de algoritmos que es más probable que utilicen son las redes neuronales (introducidas en el Capítulo 10), que vienen con su propia selección de bibliotecas avanzadas de aprendizaje automático.

Si bien Scikit-learn ofrece una variedad de algoritmos superficiales populares, TensorFlow es la biblioteca de aprendizaje automático elegida para el aprendizaje profundo/redes neuronales, ya que soporta numerosas técnicas avanzadas, incluido el cálculo automático para retropropagación/descenso de gradiente. Debido a la profundidad de los recursos, la documentación y los trabajos disponibles con TensorFlow, es el marco obvio para aprender hoy en día.

Las populares bibliotecas de redes neuronales alternativas incluyen Torch, Caffe y la creciente Keras. Escrita en Python, Keras es una biblioteca de aprendizaje profundo de código abierto que se ejecuta sobre TensorFlow, Theano y otros marcos, y permite a los usuarios realizar experimentos rápidos con menos líneas de código. Al igual que un tema de sitio web de WordPress, Keras es minimalista, modular y rápido de configurar y usar, pero es menos flexible en comparación con TensorFlow y otras bibliotecas. Los usuarios a veces utilizan Keras para validar su modelo antes de cambiar a TensorFlow para construir un modelo más personalizado.

Caffe también es de código abierto y se utiliza comúnmente para desarrollar arquitecturas de aprendizaje profundo para clasificación de imágenes y segmentación de imágenes. Caffe está escrito en C++, pero tiene una interfaz Python que también admite aceleración basada en GPU utilizando Nvidia CuDNN.

Liberado en 2002, Torch está bien establecido en la comunidad de aprendizaje profundo. Es de código abierto y está basado en el lenguaje de programación Lua. Torch ofrece una variedad de algoritmos para el aprendizaje profundo y se utiliza dentro de Facebook, Google, Twitter, NYU, IDIAP, Purdue, así como en otras compañías y laboratorios de investigación.

Hasta hace poco, Theano era otro competidor de TensorFlow, pero a fines de 2017, las contribuciones al marco se han detenido oficialmente.

A veces, junto con las redes neuronales, se utiliza otro enfoque avanzado llamado modelado de conjuntos. Esta técnica combina esencialmente algoritmos y técnicas estadísticas para crear un modelo unificado, que exploraremos más en el Capítulo 12.

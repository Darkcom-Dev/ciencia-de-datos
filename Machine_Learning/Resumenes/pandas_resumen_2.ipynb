```python
# Capitulo 15

# Agrupando datos
import pandas as pd
import numpy as np

df = pd.DataFrame({'A': ['a', 'b', 'c', 'a', 'b', 'b'],
                    'B': [2, 8, 1, 4, 3, 8],
                    'C': [102, 98, 107, 104, 115, 87]})
df.describe()
```

```python
df.groupby(['A','B']).agg(['count','mean','std'])
df

# df.agg sirve para varias operaciones al dataframe, en lugar de una sola como df.mean() o df.count()
# tambien puedes usar funciones propias que personalises.

# def my_func(x):
#     return x.max() - x.min()
# df.agg(my_func)
```

```python
np.random.seed(0)
df = pd.DataFrame({'Age': np.random.randint(20, 70, 100),
'Sex': np.random.choice(['Male', 'Female'], 100),
'number_of_foo': np.random.randint(1, 20, 100)})
df.head()
```

```python
pd.cut(df['Age'], bins=4)

# pd.cut() divide los datos en partes iguales, en este caso 4
# pd.cut() tambien puede dividir los datos en rangos a traves de una lista
```

```python
pd.cut(df['Age'], bins=[19, 40, 65, np.inf])
```

```python
age_group = pd.cut(df['Age'], bins=[19, 40, 65, np.inf])
df.groupby(age_group)['number_of_foo'].mean()
```

```python
type(age_group)
```

```python
pd.crosstab(age_group, df['Sex'])

# pd.crosstab() crea una tabla de frecuencias cruzada entre 2 o mas columnas
# la suma de las frecuencias debe dar el 100% de los datos.
```

```python
df = pd.DataFrame({'group1' : ['A', 'A', 'A', 'A','B', 'B', 'B', 'B'],
                    'group2' :['C', 'C', 'C', 'D','E', 'E', 'F', 'F'],
                    'B' :['one', np.NaN, np.NaN, np.NaN, np.NaN, 'two', np.NaN, np.NaN],
                    'C' :[np.NaN, 1, np.NaN, np.NaN,np.NaN, np.NaN, np.NaN, 4]})
                    
df
```

```python
df.fillna(0)
```

```python
#df.fillna(method='pad')
df.ffill()
# este es aquivalente a method='pad'
# ffill: propagate last valid observation forward to next valid
```

```python
#df.fillna(method='bfill') es quivalente a bfill()
df.bfill() # esto es backward fill, rellena hacia atras
```

```python


# Generar una lista de 1000 números aleatorios utilizando numpy.random.poisson
lam = 3  # Parámetro lambda de la distribución de Poisson
size = 1000  # Tamaño de la lista de números aleatorios
random_numbers = np.random.poisson(lam, size)

# Crear un gráfico de histograma
plt.hist(random_numbers, bins=20, density=True, alpha=0.7)
plt.xlabel('Número aleatorio')
plt.ylabel('Frecuencia')
plt.title('Histograma de números aleatorios de una distribución de Poisson')
plt.show()
```

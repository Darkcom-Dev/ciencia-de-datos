# Geometria Vectorial

En matemáticas, especialmente en el ámbito de la geometría y el álgebra lineal, los vectores pueden clasificarse de diversas formas según sus propiedades y características. Aquí tienes una descripción de cada uno de los términos que mencionaste:

1. **Vectores fijos**: Los vectores fijos, también conocidos como vectores posicionales o vectores de posición, son vectores que tienen un punto de origen fijo en el espacio. Se definen por su magnitud y dirección relativa a un punto de referencia. Por ejemplo, en el espacio tridimensional, un vector fijo podría representar la posición de un objeto desde el origen hasta un punto específico en el espacio.

2. **Vectores libres**: Los vectores libres, también conocidos como vectores deslizantes o trasladables, son vectores que no tienen un punto de origen fijo. Se definen por su magnitud y dirección, pero pueden ubicarse en cualquier lugar del espacio sin afectar sus propiedades. En otras palabras, dos vectores con la misma magnitud y dirección son considerados iguales, independientemente de su ubicación en el espacio. Por ejemplo, en física, un vector de fuerza se considera un vector libre, ya que su efecto no depende de su posición en el espacio.

3. **Vectores del plano**: Los vectores del plano son vectores que se encuentran dentro de un plano bidimensional. Estos vectores pueden ser fijos o libres, dependiendo de cómo se definan. En el plano cartesiano, por ejemplo, un vector fijo podría representar el desplazamiento desde el origen hasta un punto específico en el plano, mientras que un vector libre podría representar cualquier desplazamiento dentro del plano, sin estar sujeto a un punto de origen específico.

En resumen, los vectores fijos tienen un punto de origen fijo y se definen por su posición relativa a ese punto, los vectores libres pueden desplazarse libremente por el espacio sin cambiar sus propiedades esenciales, y los vectores del plano se encuentran dentro de un plano bidimensional y pueden ser fijos o libres según su definición.

## Vectores y espacios vectoriales.

Los temas centrales de vectores y espacios vectoriales incluyen:

1. **Definición de Vectores:**
   En álgebra lineal, un vector es un objeto que tiene magnitud y dirección. Puede representarse geométricamente como una flecha en el espacio tridimensional o como un conjunto ordenado de números (componentes).

2. **Operaciones con Vectores:**
   Las operaciones comunes con vectores incluyen la suma y la multiplicación por un escalar. La suma de dos vectores produce otro vector cuyas componentes son la suma de las componentes correspondientes de los vectores originales. La multiplicación por un escalar simplemente escala el vector multiplicando cada componente por el escalar.

3. **Espacios Vectoriales:**
   Un espacio vectorial es un conjunto de vectores que cumple ciertas propiedades, como la cerradura bajo la suma y la multiplicación por escalares. Estas propiedades incluyen la existencia de un vector cero, la existencia de inversos aditivos y la distributividad de la multiplicación por escalares sobre la suma de vectores.

4. **Independencia Lineal:**
   Un conjunto de vectores es linealmente independiente si ninguno de los vectores puede expresarse como una combinación lineal de los demás. Por el contrario, un conjunto de vectores es linealmente dependiente si al menos uno de los vectores puede expresarse como una combinación lineal de los demás.

5. **Bases y Dimensión:**
   Una base es un conjunto de vectores que genera todo el espacio vectorial y es linealmente independiente. La dimensión de un espacio vectorial es el número de vectores en una base. Todos los espacios vectoriales de la misma dimensión son isomorfos.

6. **Transformaciones Lineales:**
   Una transformación lineal es una función que preserva la estructura vectorial, es decir, conserva la suma de vectores y la multiplicación por escalares. Las transformaciones lineales son fundamentales en el álgebra lineal y tienen muchas aplicaciones en áreas como la geometría, la física y la ingeniería.

Estos son algunos de los temas centrales de vectores y espacios vectoriales en álgebra lineal. Comprender estos conceptos es fundamental para el estudio de áreas más avanzadas de las matemáticas, así como para su aplicación en campos como la física, la ingeniería y la informática.

## 1. Definicion de Vectores.

La definición de vectores varía según el contexto en el que se utilicen, pero en álgebra lineal y geometría, un vector es una entidad matemática que tiene magnitud y dirección. Los vectores son utilizados para representar cantidades que tienen una dirección específica en el espacio, como la velocidad, la fuerza o el desplazamiento.

Formalmente, un vector se puede representar como un conjunto ordenado de números (también llamados componentes o coordenadas) que describen su posición relativa en un sistema de coordenadas. Por ejemplo, en un sistema de coordenadas bidimensional (como el plano cartesiano), un vector puede estar representado por dos números: su componente en el eje x (horizontal) y su componente en el eje y (vertical). En un sistema de coordenadas tridimensional, un vector puede estar representado por tres números: sus componentes en los ejes x, y y z.

Geométricamente, un vector se puede representar como una flecha que parte de un punto de origen y se extiende hasta otro punto en el espacio. La longitud de la flecha representa la magnitud del vector, mientras que la dirección en la que apunta la flecha representa la dirección del vector.

Los vectores pueden ser sumados, multiplicados por escalares y sometidos a otras operaciones matemáticas para realizar diversas manipulaciones y cálculos. También pueden ser utilizados para definir rectas, planos y otros objetos geométricos, así como para representar movimientos, trayectorias y fuerzas en física.

En resumen, un vector es una entidad matemática que tiene magnitud y dirección, y que puede representarse mediante un conjunto ordenado de números en un sistema de coordenadas. Los vectores son fundamentales en álgebra lineal, geometría, física, ingeniería y otras disciplinas, donde se utilizan para representar y manipular cantidades con dirección y magnitud.

```python
class vector2d:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return vector2d(self.x + other.x, self.y + other.y)
    def __sub__(self, other):
        return vector2d(self.x - other.x, self.y - other.y)
    def __mul__(self, other):
        if isinstance(other, vector2d): # dot product
            return (self.x * other.x + self.y * other.y)
        elif isinstance(other, int) or isinstance(other, float): # scalar product
            return vector2d(self.x * other, self.y * other)
        
    def __neg__(self):
        return vector2d(-self.x, -self.y)
    def cross(self, other):
        return self.x * other.y - self.y * other.x
    def __str__(self):
        return '({}, {})'.format(self.x, self.y)
```

## Operaciones de Vectores.

Las operaciones de vectores son manipulaciones matemáticas que se realizan entre vectores para combinarlos, transformarlos o calcular propiedades relacionadas con ellos. Algunas de las operaciones de vectores más comunes incluyen:

1. **Suma de Vectores:**
   La suma de vectores es una operación que combina dos o más vectores para producir un nuevo vector, llamado vector suma. En la suma de vectores, se suman componente a componente los vectores que se están combinando. Por ejemplo, si tienes dos vectores $\vec{v}$ y $\vec{w}$ con componentes $v_x, v_y$ y $w_x, w_y$ respectivamente, entonces la suma de vectores se calcula como:
   $$
   \vec{v} + \vec{w} = (v_x + w_x, v_y + w_y)
   $$

2. **Resta de Vectores:**
   La resta de vectores es similar a la suma de vectores, pero se realiza restando componente a componente los vectores que se están combinando. Por ejemplo, la resta de $\vec{v}$ y $\vec{w}$ se calcula como:
   $$
   \vec{v} - \vec{w} = (v_x - w_x, v_y - w_y)
   $$

3. **Multiplicación Escalar:**
   La multiplicación escalar es una operación que escala un vector por un factor constante (escalar). Esto significa que cada componente del vector se multiplica por el escalar. Por ejemplo, si tienes un vector $\vec{v}$ con componentes $v_x, v_y$, y un escalar $c$, entonces la multiplicación escalar se calcula como:
   $$
   c \cdot \vec{v} = (c \cdot v_x, c \cdot v_y)
   $$

4. **Producto Punto (Producto Escalar):**
   El producto punto es una operación que toma dos vectores y produce un escalar como resultado. Se calcula sumando el producto de las componentes correspondientes de los vectores. Si tienes dos vectores $\vec{v}$ y $\vec{w}$ con componentes $v_x, v_y$ y $w_x, w_y$ respectivamente, entonces el producto punto se calcula como:
   $$
   \vec{v} \cdot \vec{w} = v_x \cdot w_x + v_y \cdot w_y
   $$

5. **Producto Cruz (Producto Vectorial):**
   El producto cruz es una operación que toma dos vectores en un espacio tridimensional y produce un nuevo vector que es perpendicular a los vectores originales. El producto cruz se utiliza comúnmente en geometría y física para calcular áreas, volúmenes y momentos. Su fórmula depende de las componentes de los vectores involucrados.

Estas son algunas de las operaciones fundamentales que se realizan entre vectores en matemáticas y ciencias. Son esenciales para realizar cálculos y resolver problemas en álgebra lineal, geometría, física y otras disciplinas.

### Suma de vectores

La suma de vectores es una operación fundamental en la que dos o más vectores se combinan para producir un nuevo vector llamado vector suma. Geométricamente, la suma de vectores se interpreta como la unión de las flechas representativas de los vectores originales, donde la flecha resultante apunta desde el punto inicial del primer vector hasta el punto final del último vector.

Matemáticamente, la suma de vectores se realiza componente a componente. Si tienes dos vectores $\vec{v}$ y $\vec{w}$ con componentes $v_x, v_y$ y $w_x, w_y$ respectivamente, entonces la suma de vectores se calcula sumando las componentes correspondientes:
   
$$
\vec{v} + \vec{w} = (v_x + w_x, v_y + w_y)
$$

Si hay más de dos vectores que se desean sumar, el proceso es el mismo: se suman las componentes correspondientes de todos los vectores para calcular las componentes del vector suma.

La suma de vectores cumple con la ley conmutativa, es decir, el orden en el que se suman los vectores no afecta al resultado final. Además, la suma de vectores cumple con la ley asociativa, lo que significa que el agrupamiento de los vectores que se suman no afecta al resultado final.

En resumen, la suma de vectores es una operación matemática y geométrica fundamental que combina dos o más vectores para producir un nuevo vector. Se realiza componente a componente y puede interpretarse como la unión de las flechas representativas de los vectores originales. Es esencial en diversas áreas, como la física, la ingeniería y la geometría, donde se utilizan para representar y calcular fuerzas, velocidades, desplazamientos y otras cantidades vectoriales.

### Multiplicación Escalar de Vectores.

La multiplicación escalar de un vector es una operación en la que se escala el vector por un factor constante, conocido como escalar. Esto significa que cada componente del vector se multiplica por el escalar, resultando en un nuevo vector que tiene la misma dirección que el vector original pero una magnitud diferente.

Formalmente, si tienes un vector $\vec{v}$ con componentes $v_x$ y $v_y$, y un escalar $c$, entonces la multiplicación escalar se calcula como:

$$
c \cdot \vec{v} = (c \cdot v_x, c \cdot v_y)
$$

Por ejemplo, si tienes el vector $\vec{v} = (2, -3)$ y quieres multiplicarlo por el escalar $3$, el resultado sería:

$$
3 \cdot \vec{v} = (3 \cdot 2, 3 \cdot (-3)) = (6, -9)
$$

Geométricamente, la multiplicación escalar de un vector por un número positivo aumenta su magnitud, mientras que la multiplicación por un número negativo invierte su dirección. Si el escalar es igual a $1$, el vector resultante es el mismo que el vector original, ya que no se produce ningún cambio en la magnitud. Si el escalar es igual a $0$, el vector resultante es el vector nulo, ya que todos sus componentes se vuelven cero.

La multiplicación escalar de vectores es una operación importante en álgebra lineal y se utiliza en una variedad de aplicaciones, como la física, la ingeniería, la informática y la geometría. Se emplea para escalar cantidades vectoriales, como fuerzas, velocidades, desplazamientos, y en operaciones de proyección y escalado en gráficos por computadora, entre otras aplicaciones.

### Producto Punto.

El producto punto, también conocido como producto escalar, es una operación entre dos vectores que produce un escalar como resultado. El producto punto se calcula multiplicando componente a componente los dos vectores y luego sumando los productos resultantes. Formalmente, si tienes dos vectores $\vec{v} = (v_1, v_2, \ldots, v_n)$ y $\vec{w} = (w_1, w_2, \ldots, w_n)$, entonces el producto punto se calcula como:

$$
\vec{v} \cdot \vec{w} = v_1 \cdot w_1 + v_2 \cdot w_2 + \ldots + v_n \cdot w_n
$$

Geométricamente, el producto punto representa la proyección de un vector sobre otro. Si los dos vectores están en la misma dirección, el producto punto será positivo, y si están en direcciones opuestas, el producto punto será negativo. Si los dos vectores son ortogonales (perpendiculares), el producto punto será cero.

El producto punto tiene varias propiedades importantes:

1. **Conmutatividad:** $\vec{v} \cdot \vec{w} = \vec{w} \cdot \vec{v}$
2. **Distributividad respecto a la suma:** $\vec{v} \cdot (\vec{w} + \vec{u}) = \vec{v} \cdot \vec{w} + \vec{v} \cdot \vec{u}$
3. **Multiplicación por un escalar:** $(c \cdot \vec{v}) \cdot \vec{w} = c \cdot (\vec{v} \cdot \vec{w}) = \vec{v} \cdot (c \cdot \vec{w})$
4. **Linealidad:** $(\vec{v} + \vec{w}) \cdot \vec{u} = \vec{v} \cdot \vec{u} + \vec{w} \cdot \vec{u}$

El producto punto tiene numerosas aplicaciones en física, geometría, ingeniería y otras disciplinas. Se utiliza para calcular la proyección de un vector sobre otro, encontrar el ángulo entre dos vectores, determinar si dos vectores son ortogonales, resolver sistemas de ecuaciones lineales y en el análisis de señales, entre otras aplicaciones.

### Producto cruz.

El producto cruz, también conocido como producto vectorial, es una operación entre dos vectores que produce un nuevo vector perpendicular a los vectores originales. El producto cruz se denota comúnmente como $\vec{v} \times \vec{w}$, donde $\vec{v}$ y $\vec{w}$ son los vectores de entrada.

La fórmula para calcular el producto cruz entre dos vectores tridimensionales $\vec{v} = (v_1, v_2, v_3)$ y $\vec{w} = (w_1, w_2, w_3)$ es la siguiente:

$$
\vec{v} \times \vec{w} = \left( v_2 w_3 - v_3 w_2, v_3 w_1 - v_1 w_3, v_1 w_2 - v_2 w_1 \right)
$$

Este cálculo produce un nuevo vector cuyas componentes son combinaciones lineales de las componentes de los vectores originales. El vector resultante es perpendicular al plano definido por los vectores de entrada y su magnitud es proporcional al área del paralelogramo definido por los vectores originales.

El producto cruz tiene varias propiedades importantes:

1. **Anticonmutatividad:** $\vec{v} \times \vec{w} = -(\vec{w} \times \vec{v})$
2. **Distributividad:** $\vec{v} \times (\vec{w} + \vec{u}) = (\vec{v} \times \vec{w}) + (\vec{v} \times \vec{u})$
3. **Multiplicación por un escalar:** $c \cdot (\vec{v} \times \vec{w}) = (c \cdot \vec{v}) \times \vec{w} = \vec{v} \times (c \cdot \vec{w})$
4. **No asociatividad:** En general, $(\vec{v} \times \vec{w}) \times \vec{u} \neq \vec{v} \times (\vec{w} \times \vec{u})$

El producto cruz se utiliza en física y geometría para calcular momentos, momentos angulares, áreas de paralelogramos y volúmenes de paralelepípedos. También es fundamental en la formulación de leyes físicas, como las leyes de Maxwell en electromagnetismo y las leyes de movimiento de Newton en la mecánica.

# Matrices y Operaciones de Matrices

Los temas centrales de matrices y operaciones con matrices incluyen:

1. **Definición de Matrices:**
   Una matriz es un arreglo rectangular de números dispuestos en filas y columnas. Cada número en una matriz se llama elemento. La dimensión de una matriz se describe por el número de filas y columnas que contiene.

2. **Operaciones Básicas con Matrices:**
   Las operaciones básicas con matrices incluyen la suma y la resta de matrices, que se realizan sumando o restando los elementos correspondientes de las matrices. La multiplicación por un escalar es otra operación común que implica multiplicar cada elemento de la matriz por un número escalar.

3. **Multiplicación de Matrices:**
   La multiplicación de matrices es una operación fundamental en álgebra lineal. La multiplicación de dos matrices se realiza multiplicando las filas de la primera matriz por las columnas de la segunda matriz y sumando los productos resultantes. Es importante tener en cuenta que no todas las matrices se pueden multiplicar entre sí, ya que deben cumplir ciertas condiciones de compatibilidad.

4. **Matriz Transpuesta:**
   La transposición de una matriz implica intercambiar sus filas y columnas. Esto se realiza colocando los elementos de la fila i en la columna i y viceversa. La matriz transpuesta se denota comúnmente con el símbolo "T".

5. **Determinantes:**
   El determinante es un valor numérico asociado a una matriz cuadrada. Se utiliza en una variedad de aplicaciones, incluyendo la resolución de sistemas de ecuaciones lineales, el cálculo de la inversa de una matriz y la resolución de sistemas de ecuaciones diferenciales lineales.

6. **Inversas de Matrices:**
   La inversa de una matriz es otra matriz que, cuando se multiplica por la matriz original, produce la matriz identidad. No todas las matrices tienen inversas, y las que sí tienen inversas se denominan matrices invertibles o no singulares.

7. **Sistemas de Ecuaciones Lineales:**
   Las matrices se utilizan para representar sistemas de ecuaciones lineales de forma compacta y eficiente. Resolver un sistema de ecuaciones lineales es equivalente a encontrar la solución de un sistema de ecuaciones matriciales.

Estos son algunos de los temas centrales de matrices y operaciones con matrices en álgebra lineal. Comprender estos conceptos es fundamental para su aplicación en una variedad de campos, incluyendo la física, la ingeniería, la informática y la economía.

### 1. Definición de Matrices

En álgebra lineal, una matriz es una colección rectangular de números o elementos dispuestos en filas y columnas. Formalmente, una matriz $ m \times n $ es una disposición rectangular de $ m $ filas y $ n $ columnas, donde cada elemento de la matriz está identificado por su posición en la fila y la columna.

Por ejemplo, una matriz $ 2 \times 3 $ se vería así:

$$
\begin{pmatrix}
a & b & c \\
d & e & f \\
\end{pmatrix}
$$

Donde $ a, b, c, d, e, f $ son los elementos individuales de la matriz.

Las matrices se utilizan para representar datos tabulares, sistemas de ecuaciones lineales, transformaciones lineales y muchas otras aplicaciones en matemáticas y ciencias aplicadas. Las operaciones básicas con matrices incluyen la suma de matrices, la multiplicación de matrices, la transposición de matrices y la inversión de matrices, entre otras. Las matrices son una herramienta fundamental en el estudio del álgebra lineal y tienen numerosas aplicaciones en diversos campos, como la física, la ingeniería, la informática, la economía y la estadística.

```python
class matrix2x2:
    def __init__(self, x, y, z, w):
        self.x = x
        self.y = y
        self.z = z
        self.w = w
    def __add__(self, other):
        return matrix2x2(self.x + other.x, self.y + other.y, self.z + other.z, self.w + other.w)
    def __sub__(self, other):
        return matrix2x2(self.x - other.x, self.y - other.y, self.z - other.z, self.w - other.w)
    def __mul__(self, other):
        if isinstance(other, matrix2x2):
            return matrix2x2(self.x * other.x + self.y * other.z, self.x * other.y + self.y * other.w,
                             self.z * other.x + self.w * other.z, self.z * other.y + self.w * other.w)
        elif isinstance(other, int) or isinstance(other, float):
            return matrix2x2(self.x * other, self.y * other, self.z * other, self.w * other)
    def __neg__(self):
        return matrix2x2(-self.x, -self.y, -self.z, -self.w)
    def __str__(self):
        return '({}, {})\n, ({}, {})'.format(self.x, self.y, self.z, self.w)
```

```python
class matrix3x3:
    def __init__(self, x, y, z, w, u, v, r, s, t):
        self.x = x
        self.y = y
        self.z = z
        self.w = w
        self.u = u
        self.v = v
        self.r = r
        self.s = s
        self.t = t

    def __add__(self, other):
        return matrix3x3(self.x + other.x, self.y + other.y, self.z + other.z, self.w + other.w,
                         self.u + other.u, self.v + other.v, self.r + other.r, self.s + other.s, self.t + other.t)
    def __sub__(self, other):
        return matrix3x3(self.x - other.x, self.y - other.y, self.z - other.z, self.w - other.w,
                         self.u - other.u, self.v - other.v, self.r - other.r, self.s - other.s, self.t - other.t)
    def __mul__(self, other):
        if isinstance(other, matrix3x3):
            return matrix3x3(self.x * other.x + self.y * other.u + self.z * other.r, self.x * other.y + self.y * other.v + self.z * other.s, self.x * other.z + self.y * other.w + self.z * other.t,
                             self.w * other.x + self.v * other.u + self.t * other.r, self.w * other.y + self.v * other.v + self.t * other.s, self.w * other.z + self.v * other.w + self.t * other.t,
                             self.u * other.x + self.s * other.u + self.r * other.r, self.u * other.y + self.s * other.v + self.r * other.s, self.u * other.z + self.s * other.w + self.r * other.t)
        elif isinstance(other, int) or isinstance(other, float):
            return matrix3x3(self.x * other, self.y * other, self.z * other, self.w * other, self.u * other, self.v * other, self.r * other, self.s * other, self.t * other)
        
    def __neg__(self):
        return matrix3x3(-self.x, -self.y, -self.z, -self.w, -self.u, -self.v, -self.r, -self.s, -self.t)
    
    def __str__(self):
        return '({}, {}, {})\n, ({}, {}, {})\n, ({}, {}, {})'.format(self.x, self.y, self.z, self.w, self.u, self.v, self.r, self.s, self.t)
```

# 2. Operaciones Básicas de Matrices.

Claro, aquí tienes algunos ejemplos de operaciones básicas con matrices:

1. **Suma de matrices:** Dadas dos matrices $ A $ y $ B $ del mismo tamaño $ m \times n $, la suma de matrices se realiza sumando los elementos correspondientes de ambas matrices. Por ejemplo:

$$
\begin{pmatrix}
1 & 2 \\
3 & 4 \\
\end{pmatrix}
+
\begin{pmatrix}
5 & 6 \\
7 & 8 \\
\end{pmatrix}
=
\begin{pmatrix}
1+5 & 2+6 \\
3+7 & 4+8 \\
\end{pmatrix}
=
\begin{pmatrix}
6 & 8 \\
10 & 12 \\
\end{pmatrix}
$$

2. **Multiplicación de una matriz por un escalar:** Dada una matriz $ A $ y un escalar $ k $, la multiplicación de la matriz por el escalar se realiza multiplicando cada elemento de la matriz por el escalar. Por ejemplo:

$$
3 \times
\begin{pmatrix}
1 & 2 \\
3 & 4 \\
\end{pmatrix}
=
\begin{pmatrix}
3 \times 1 & 3 \times 2 \\
3 \times 3 & 3 \times 4 \\
\end{pmatrix}
=
\begin{pmatrix}

\end{pmatrix}
$$

3. **Multiplicación de matrices:** Dadas dos matrices $ A $ de tamaño $ m \times p $ y $ B $ de tamaño $ p \times n $, la multiplicación de matrices se realiza multiplicando cada fila de la primera matriz por cada columna de la segunda matriz y sumando los productos resultantes. Por ejemplo:

$$
\begin{pmatrix}
1 & 2 \\
3 & 4 \\
\end{pmatrix}
\times
\begin{pmatrix}
5 & 6 \\
7 & 8 \\
\end{pmatrix}
=
\begin{pmatrix}
(1 \times 5 + 2 \times 7) & (1 \times 6 + 2 \times 8) \\
(3 \times 5 + 4 \times 7) & (3 \times 6 + 4 \times 8) \\
\end{pmatrix}
=
\begin{pmatrix}
19 & 22 \\
43 & 50 \\
\end{pmatrix}
$$

Estos son solo algunos ejemplos de las operaciones básicas que se pueden realizar con matrices. Las matrices y sus operaciones son fundamentales en el álgebra lineal y tienen numerosas aplicaciones en matemáticas y ciencias aplicadas.

# Determinantes y Sistemas de ecuaciones lineales.

Los temas centrales de determinantes y sistemas de ecuaciones lineales incluyen:

1. **Determinantes:**
   - Definición: El determinante es un valor numérico asociado a una matriz cuadrada.
   - Propiedades: Los determinantes tienen propiedades útiles, como la multiplicatividad por escalares, la suma de matrices y la transposición.
   - Cálculo: Hay varios métodos para calcular determinantes, como el método de eliminación de Gauss, el método de cofactores y el método de triangularización.

2. **Sistemas de Ecuaciones Lineales:**
   - Definición: Un sistema de ecuaciones lineales consiste en un conjunto de ecuaciones lineales que comparten las mismas variables.
   - Formas: Los sistemas de ecuaciones lineales pueden presentarse en forma de matriz (forma matricial) o en forma de ecuaciones algebraicas.
   - Soluciones: Un sistema de ecuaciones lineales puede tener una solución única, infinitas soluciones o ninguna solución. La naturaleza de las soluciones se determina por el número de ecuaciones y variables y la relación entre ellas.
   - Métodos de resolución: Los métodos comunes para resolver sistemas de ecuaciones lineales incluyen el método de eliminación de Gauss, el método de sustitución, el método de matrices aumentadas y el método de inversión de matrices.

3. **Regla de Cramer:**
   - La regla de Cramer es un método para resolver sistemas de ecuaciones lineales utilizando determinantes.
   - Requiere que la matriz de coeficientes del sistema sea invertible y que haya tantas ecuaciones como variables.

4. **Aplicaciones:**
   - Los sistemas de ecuaciones lineales tienen una amplia variedad de aplicaciones en ciencias e ingeniería, incluyendo el análisis de circuitos eléctricos, la resolución de problemas de optimización, la modelización de fenómenos físicos y la economía.

Estos son los temas centrales de determinantes y sistemas de ecuaciones lineales en álgebra lineal. Comprender estos conceptos es esencial para resolver problemas prácticos en una variedad de disciplinas y aplicaciones.

# Espacios Vectoriales de dimensión finita y bases.

Los temas centrales de espacios vectoriales de dimensión finita y bases incluyen:

1. **Espacios Vectoriales:**
   - Definición: Un espacio vectorial es un conjunto de vectores sobre un campo que cumple ciertas propiedades, como la cerradura bajo la suma y la multiplicación por escalares.
   - Propiedades: Los espacios vectoriales deben satisfacer propiedades como la existencia de un vector cero, la existencia de inversos aditivos, y la cerradura bajo la suma y la multiplicación por escalares.

2. **Bases:**
   - Definición: Una base de un espacio vectorial es un conjunto de vectores que genera todo el espacio y es linealmente independiente.
   - Propiedades: Una base es un conjunto mínimo de vectores necesarios para expresar cualquier vector en el espacio vectorial como una combinación lineal de los vectores de la base.
   - Dimensión: La dimensión de un espacio vectorial es el número de vectores en una base. Todos los espacios vectoriales de la misma dimensión son isomorfos.

3. **Coordenadas y Representación Matricial:**
   - Coordenadas: Cada vector en un espacio vectorial puede ser representado de manera única como una combinación lineal de los vectores de la base.
   - Matrices: Los vectores de la base pueden ser organizados como columnas en una matriz, llamada matriz de cambio de base, que permite transformar las coordenadas de un vector de una base a otra.

4. **Subespacios Vectoriales:**
   - Definición: Un subespacio vectorial de un espacio vectorial V es un conjunto de vectores en V que es cerrado bajo la suma y la multiplicación por escalares.
   - Propiedades: Los subespacios vectoriales también son espacios vectoriales y heredan muchas de las propiedades de V.

5. **Dependencia e Independencia Lineal:**
   - Dependencia Lineal: Un conjunto de vectores es linealmente dependiente si al menos uno de los vectores puede expresarse como una combinación lineal de los demás.
   - Independencia Lineal: Un conjunto de vectores es linealmente independiente si ninguno de los vectores puede expresarse como una combinación lineal de los demás.

6. **Aplicaciones:**
   - Las bases y los espacios vectoriales de dimensión finita tienen aplicaciones en una variedad de áreas, incluyendo la física, la ingeniería, la economía y la informática. Se utilizan para modelar fenómenos físicos, resolver sistemas de ecuaciones lineales, y en la optimización y el análisis de datos, entre otros.

Estos son los temas centrales de espacios vectoriales de dimensión finita y bases en álgebra lineal. Comprender estos conceptos es fundamental para el estudio y la aplicación de álgebra lineal en una variedad de disciplinas.

# Transformaciones lineales.

Los temas centrales de transformaciones lineales incluyen:

1. **Definición:**
   - Una transformación lineal es una función entre dos espacios vectoriales que preserva la estructura vectorial, es decir, conserva la suma de vectores y la multiplicación por escalares.

2. **Propiedades:**
   - Las transformaciones lineales tienen varias propiedades importantes, como la preservación de la suma y la multiplicación por escalares.
   - Son lineales, lo que significa que T(u + v) = T(u) + T(v) y T(cu) = cT(u) para todos los vectores u y v y todos los escalares c.

3. **Matriz de Transformación:**
   - Cada transformación lineal entre dos espacios vectoriales puede representarse mediante una matriz.
   - La matriz de transformación puede calcularse aplicando la transformación a los vectores de la base y organizando los resultados en columnas.

4. **Núcleo y Rango:**
   - El núcleo (o kernel) de una transformación lineal es el conjunto de todos los vectores que se mapean al vector cero en el espacio de llegada.
   - El rango de una transformación lineal es la dimensión del espacio generado por las imágenes de todos los vectores del espacio de partida.

5. **Inyectividad y Sobreyectividad:**
   - Una transformación lineal es inyectiva si cada vector en el espacio de partida tiene una imagen única en el espacio de llegada.
   - Una transformación lineal es sobreyectiva si cada vector en el espacio de llegada tiene al menos un preimagen en el espacio de partida.

6. **Valores y Vectores Propios:**
   - Los valores propios (o autovalores) y los vectores propios (o autovectores) son conceptos importantes en transformaciones lineales.
   - Un vector propio de una transformación lineal es un vector que se mapea a un múltiplo escalar de sí mismo bajo la transformación.

7. **Diagonalización:**
   - Algunas transformaciones lineales pueden diagonalizarse, lo que significa que pueden ser representadas por una matriz diagonal.
   - La diagonalización de una matriz de transformación facilita el cálculo de potencias de la matriz y la resolución de sistemas de ecuaciones diferenciales lineales.

8. **Aplicaciones:**
   - Las transformaciones lineales tienen una amplia variedad de aplicaciones en campos como la geometría, la física, la ingeniería y la informática.
   - Se utilizan para modelar fenómenos físicos, resolver sistemas de ecuaciones lineales, comprimir imágenes y datos, y en la programación de gráficos por computadora, entre otros.

Estos son los temas centrales de las transformaciones lineales en álgebra lineal. Comprender estos conceptos es esencial para su aplicación en una variedad de disciplinas y aplicaciones.

# Valores propios y Vectores Propios.

Los temas centrales de valores propios y vectores propios incluyen:

1. **Definición:**
   - Los valores propios (o autovalores) y los vectores propios (o autovectores) son conceptos fundamentales en álgebra lineal.
   - Un vector propio de una transformación lineal es un vector que, después de ser transformado por la matriz de la transformación, se mantiene en la misma dirección (o en la dirección opuesta) que el vector original.
   - Un valor propio es un escalar λ tal que existe un vector no nulo v para el cual la multiplicación de la matriz de transformación A por v es igual a λ veces v.

2. **Cálculo de Valores Propios y Vectores Propios:**
   - Para encontrar los valores propios y vectores propios de una matriz, se resuelve la ecuación característica det(A - λI) = 0, donde A es la matriz de transformación, λ es el valor propio y I es la matriz identidad.
   - Una vez que se encuentran los valores propios, los correspondientes vectores propios se calculan resolviendo el sistema de ecuaciones (A - λI)v = 0.

3. **Diagonalización:**
   - Algunas matrices pueden diagonalizarse, lo que significa que pueden ser descompuestas en una matriz diagonal formada por sus valores propios y una matriz de vectores propios.
   - La diagonalización facilita el cálculo de potencias de la matriz y la resolución de sistemas de ecuaciones diferenciales lineales.

4. **Aplicaciones:**
   - Los valores propios y vectores propios tienen numerosas aplicaciones en álgebra lineal y en diferentes campos.
   - Se utilizan en el análisis de sistemas dinámicos, en la resolución de ecuaciones diferenciales lineales, en la compresión de datos, en la reducción de dimensionalidad y en el análisis de estructuras y comportamientos en física, ingeniería, ciencias de la computación, economía y otros campos.

5. **Propiedades:**
   - Los valores propios pueden ser reales o complejos.
   - Si una matriz es simétrica (o hermitiana en el caso de matrices complejas), todos sus valores propios son reales y los vectores propios correspondientes son ortogonales entre sí.

Estos son los temas centrales de valores propios y vectores propios en álgebra lineal. Comprender estos conceptos es esencial para su aplicación en una variedad de disciplinas y aplicaciones.

# Espacios Vectoriales

Un espacio vectorial $ V $ sobre un campo $ F $ es un conjunto de elementos llamados vectores, junto con dos operaciones, **suma de vectores y multiplicación por un escalar**, que satisfacen ciertas propiedades.

1. **Conjunto de vectores:** Un espacio vectorial consiste en un conjunto de elementos llamados vectores. Estos vectores pueden ser representados como tuplas de números, matrices, funciones, polinomios, entre otros.

2. **Suma de vectores:** Para cada par de vectores $ \mathbf{u} $ y $ \mathbf{v} $ en $ V $, hay una operación de suma $ \mathbf{u} + \mathbf{v} $ que produce otro vector en $ V $. Esta operación debe ser conmutativa, asociativa y tener un vector cero ($ \mathbf{0} $).

3. **Multiplicación por un escalar:** Para cada vector $ \mathbf{v} $ en $ V $ y cada escalar $ c $ en $ F $, hay una operación de multiplicación $ c \mathbf{v} $ que produce otro vector en $ V $. Esta operación debe ser distributiva respecto a la suma de vectores y debe cumplir ciertas propiedades con respecto a los escalares y la identidad multiplicativa del campo $ F $.

4. **Propiedades adicionales:** Los espacios vectoriales suelen tener otras propiedades, como la existencia de vectores opuestos (para cada vector $ \mathbf{v} $ en $ V $, hay un vector $ -\mathbf{v} $ en $ V $ tal que $ \mathbf{v} + (-\mathbf{v}) = \mathbf{0} $), y la combinación lineal de vectores.

5. **Subespacios vectoriales:** Un subconjunto $ W $ de un espacio vectorial $ V $ es un subespacio vectorial de $ V $ si $ W $ también es un espacio vectorial bajo las mismas operaciones de suma de vectores y multiplicación por un escalar definidas en $ V $.

Los espacios vectoriales son una herramienta poderosa en matemáticas y tienen muchas aplicaciones en diversas áreas, incluyendo la física, la ingeniería, la informática y la economía. Estudiar sus propiedades y estructuras nos permite comprender mejor la naturaleza de los vectores y desarrollar técnicas para resolver problemas en diversas disciplinas.

# Grupo Abeliano

Un grupo abeliano, también conocido como grupo conmutativo, es un tipo especial de estructura algebraica en la que la operación binaria definida en el conjunto cumple con la propiedad conmutativa. Formalmente, un grupo abeliano se define como sigue:

Sea $ G $ un conjunto no vacío junto con una operación binaria $ \cdot $ definida en $ G $, denotada como $ (G, \cdot) $. $ G $ es un grupo abeliano si cumple con las siguientes propiedades:

1. **Cerradura:** Para todos $ a, b $ en $ G $, el resultado de la operación $ a \cdot b $ también está en $ G $.

2. **Asociatividad:** Para todos $ a, b, c $ en $ G $, la operación es asociativa, es decir, $ (a \cdot b) \cdot c = a \cdot (b \cdot c) $.

3. **Elemento neutro:** Existe un elemento en $ G $, denotado como $ e $ o $ 1 $, tal que para todo $ a $ en $ G $, $ a \cdot e = e \cdot a = a $. Este elemento neutro actúa como la identidad bajo la operación.

4. **Inverso:** Para cada elemento $ a $ en $ G $, existe un elemento $ a^{-1} $ en $ G $ tal que $ a \cdot a^{-1} = a^{-1} \cdot a = e $. Este elemento inverso de $ a $ es único.

5. **Conmutatividad:** Para todos $ a, b $ en $ G $, la operación es conmutativa, es decir, $ a \cdot b = b \cdot a $.

Un ejemplo común de un grupo abeliano es el conjunto de números enteros bajo la operación de suma ($ (\mathbb{Z}, +) $). Otros ejemplos incluyen el conjunto de números reales bajo la suma, el conjunto de matrices cuadradas invertibles bajo la multiplicación, y el conjunto de vectores en un espacio vectorial bajo la suma vectorial.

Los grupos abelianos son importantes en matemáticas y tienen numerosas aplicaciones en diferentes áreas, como la teoría de números, la geometría algebraica, la física y la criptografía.

Este espacio vectorial es un subespacio del espacio de los vectores de $ n $ dimensiones, donde $ n $ es el número de incógnitas.

El espacio de todos los vectores libres en $ \mathbb{R}^3 $ forma un espacio vectorial tridimensional.

3. **Matrices con coeficientes en un cuerpo $ K $:** Las matrices con coeficientes en un cuerpo $ K $ también forman un espacio vectorial. La suma de matrices y la multiplicación por escalares definen operaciones que cumplen con las propiedades de un espacio vectorial. Por ejemplo, el conjunto de todas las matrices $ m \times n $ con coeficientes en $ K $ forma un espacio vectorial sobre $ K $.

4. **Polinomios con coeficientes en un cuerpo $ K $:** Los polinomios con coeficientes en un cuerpo $ K $ también forman un espacio vectorial. La suma de polinomios y la multiplicación por escalares definen operaciones que cumplen con las propiedades de un espacio vectorial. Por ejemplo, el conjunto de todos los polinomios de grado $ n $ o menos con coeficientes en $ K $ forma un espacio vectorial sobre $ K $.

En resumen, todos estos objetos son ejemplos de espacios vectoriales sobre un cuerpo $ K $, donde la operación de suma y la multiplicación por escalares están definidas de manera que cumplen con las propiedades de un espacio vectorial.

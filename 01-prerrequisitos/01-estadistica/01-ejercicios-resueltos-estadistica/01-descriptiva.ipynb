Calculo de la media aritmética, la mediana y la moda. Se analizó el IVA que se aplica, en diversos países europeos, a la compra de obras de arte. Los resultados obtenidos fueron los siguientes:

PAIS

#### Solución 10.
Ahora realizamos las cuatro distribuciones de frecuencias:

Xi      ni  fi      Ni  Fi
0,06    3   0,375   3   0,375
0,07    1   0,125   4   0,500
0,16    1   0,125   5   0,625
0,17    1   0,125   6   0,750
0,20    1   0,125   7   0,875
0,22    1   0,125   8   1
__________________________
Total   8   1

Calculamos la media aritmética:

$$a = \frac{\sum{x_i-n_i}}{n}=\frac{1}{8}=0.125$$

Ahora calculamos la mediana:

$$Me = \frac{x_{j-1}+x_j}{2}=\frac{0.07+0.16}{2}=0.115$$

Por último, el valor mas frecuente, correspondiente a la moda, es el valor: $x_j = 0.06$ Por tanto: $Md = 0.06$

# SETTING UP YOUR DATA

Una vez que hayas limpiado tu conjunto de datos, el siguiente paso es dividir los datos en dos segmentos para prueba y entrenamiento. Es muy importante no probar tu modelo con los mismos datos que utilizaste para entrenarlo. La proporción de las dos divisiones debe ser aproximadamente de 70/30 o 80/20. Esto significa que tus datos de entrenamiento deben representar entre el 70 por ciento y el 80 por ciento de las filas en tu conjunto de datos, y el otro 20 por ciento a 30 por ciento de las filas es tu conjunto de datos de prueba. Es vital dividir tus datos por filas y no por columnas.

Antes de dividir tus datos, es importante que aleatorices todas las filas en el conjunto de datos. Esto ayuda a evitar sesgos en tu modelo, ya que tu conjunto de datos original podría estar organizado secuencialmente dependiendo del momento en que se recopiló o algún otro factor. A menos que aleatories tus datos, podrías omitir accidentalmente una variación importante de los datos de entrenamiento que causará sorpresas no deseadas al aplicar el modelo entrenado a tus datos de prueba. Afortunadamente, Scikit-learn proporciona una función incorporada para mezclar y aleatorizar tus datos con solo una línea de código (demonstrado en el Capítulo 13).

Después de aleatorizar tus datos, puedes comenzar a diseñar tu modelo y aplicarlo a los datos de entrenamiento. El restante 30 por ciento aproximadamente de datos se aparta y se reserva para probar la precisión del modelo.

En el caso del aprendizaje supervisado, el modelo se desarrolla alimentando a la máquina los datos de entrenamiento y la salida esperada (y). La máquina puede analizar y discernir las relaciones entre las características (X) encontradas en los datos de entrenamiento para calcular la salida final (y).

El siguiente paso es medir qué tan bien realmente se desempeña el modelo. Un enfoque común para analizar la precisión de la predicción es una medida llamada error absoluto medio, que examina cada predicción en el modelo y proporciona un puntaje de error promedio para cada predicción.

En Scikit-learn, el error absoluto medio se encuentra usando la función model.predict en X (características). Esto funciona primero conectando los valores de y del conjunto de datos de entrenamiento y generando una predicción para cada fila en el conjunto de datos. Scikit-learn comparará las predicciones del modelo con el resultado correcto y medirá su precisión. Sabrás si tu modelo es preciso cuando la tasa de error entre el conjunto de datos de entrenamiento y prueba sea baja. Esto significa que el modelo ha aprendido los patrones y tendencias subyacentes del conjunto de datos.

Una vez que el modelo pueda predecir adecuadamente los valores de los datos de prueba, está listo para su uso en la práctica. Si el modelo no puede predecir con precisión los valores de los datos de prueba, deberás verificar si los datos de entrenamiento y prueba se aleatorizaron correctamente. Alternativamente, es posible que necesites cambiar los hiperparámetros del modelo.

Cada algoritmo tiene hiperparámetros; estos son los ajustes de tu algoritmo. En términos simples, estos ajustes controlan e impactan en qué tan rápido el modelo aprende patrones y qué patrones identificar y analizar.

#### Cross Validation

Aunque la división de datos de entrenamiento/prueba puede ser efectiva para desarrollar modelos a partir de datos existentes, queda la duda sobre si el modelo funcionará con nuevos datos. Si tu conjunto de datos existente es demasiado pequeño para construir un modelo preciso, o si la partición de datos de entrenamiento/prueba no es apropiada, esto puede llevar a estimaciones deficientes del rendimiento en la práctica.

Afortunadamente, hay una solución efectiva para este problema. En lugar de dividir los datos en dos segmentos (uno para entrenamiento y otro para prueba), podemos implementar lo que se conoce como validación cruzada. La validación cruzada maximiza la disponibilidad de datos de entrenamiento dividiendo los datos en varias combinaciones y probando cada combinación específica.

La validación cruzada se puede realizar a través de dos métodos principales. El primer método es la validación cruzada exhaustiva, que implica encontrar y probar todas las combinaciones posibles para dividir la muestra original en un conjunto de entrenamiento y un conjunto de prueba. El método alternativo y más común es la validación cruzada no exhaustiva, conocida como validación k-fold. La técnica de validación k-fold consiste en dividir los datos en k grupos asignados y reservar uno de esos grupos para probar el modelo de entrenamiento en cada ronda.

Para realizar la validación k-fold, los datos se asignan primero aleatoriamente a k grupos de igual tamaño. Uno de los grupos se reserva como el grupo de prueba y se utiliza para medir y evaluar el rendimiento de los otros (k-1) grupos.

Figure 2: k-fold validation

El proceso de validación cruzada se repite k veces ("folds"). En cada fold, se reserva un grupo para probar el modelo de entrenamiento generado por los otros grupos. El proceso se repite hasta que todos los grupos se hayan utilizado tanto como grupo de entrenamiento como de prueba. Los resultados se agregan y combinan para formular un solo modelo.

Al utilizar todos los datos disponibles tanto para fines de entrenamiento como de prueba, la técnica de validación k-fold minimiza dramáticamente el error potencial (como el sobreajuste) encontrado al depender de una división fija de datos de entrenamiento y prueba.

#### How Much Data Do I Need?

Una pregunta común para los estudiantes que comienzan en el aprendizaje automático es ¿cuántos datos necesito para entrenar mi conjunto de datos? En general, el aprendizaje automático funciona mejor cuando tu conjunto de datos de entrenamiento incluye una gama completa de combinaciones de características.

¿Cómo se ve una gama completa de combinaciones de características? Imagina que tienes un conjunto de datos sobre científicos de datos categorizados por las siguientes características:
- Grado universitario (X)
- Experiencia profesional de 5+ años (X)
- Hijos (X)
- Salario (y)
  
Para evaluar la relación que las tres primeras características (X) tienen con el salario de un científico de datos (y), necesitamos un conjunto de datos que incluya el valor de y para cada combinación de características. Por ejemplo, necesitamos conocer el salario de los científicos de datos con un grado universitario, 5+ años de experiencia profesional y que no tienen hijos, así como los científicos de datos con un grado universitario, 5+ años de experiencia profesional y que sí tienen hijos.

Cuantas más combinaciones disponibles, más efectivo será el modelo para capturar cómo afecta cada atributo a y (el salario del científico de datos). Esto garantizará que cuando se ponga el modelo en práctica en los datos de prueba o en datos del mundo real, no se desmorone inmediatamente ante la vista de combinaciones no vistas.

Como mínimo, un modelo de aprendizaje automático debe tener típicamente diez veces más puntos de datos que el número total de características. Entonces, para un conjunto de datos pequeño con tres características, los datos de entrenamiento idealmente deberían tener al menos treinta filas.

El otro punto a recordar es que más datos relevantes generalmente son mejores que menos. Tener más datos relevantes te permite cubrir más combinaciones y generalmente ayuda a garantizar predicciones más precisas. En algunos casos, puede que no sea posible o rentable obtener datos para todas las combinaciones posibles. En estos casos, tendrás que conformarte con los datos que tengas a tu disposición.

### Gráficos de datos básicos.

Los usos de Pandas proporcionan múltiples formas de hacer gráficos de los datos dentro del marco de datos. Utiliza `matplotlib` para ese propósito.

Los gráficos básicos tienen sus contenedores para objetos DataFrame y Series:

#### Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(
    {
        'x': [10,8,10,7,7,10,9,9],
        'y': [6,4,5,5,7,10,9,9]
    }
)
df.plot()
```

Puede llamar al mismo método para que un objeto Serie trace un subconjunto del DataFame:

```python
df.x.plot()
```

#### Bar Chart - gráfico de barras

Si desea explorar la distribución de sus datos, puede utilizar el método `hist()`.

```python
df.x.hist()
```

#### Método general para trazar `plot()`

Todos los gráficos posibles están disponibles a través del método de trazado. El tipo de gráfico se selecciona mediante el argumento kind.

```python
df.x.plot(kind='pie')
```

> Nota: En muchos entornos, el gráfico circular tendrá un aspecto ovalado. Para convertirlo en un círculo, use lo siguiente:

```python
# en el libro se usa pyplot.axis('equal')
# pero como en la celda de arriba se importó matplotlib.pyplot as plt
# en lugar de solo matplotlib, en su lugar se usa el alias plt.axis
plt.axis('equal')
df.x.plot(kind='pie')
```

### Estilizando el trazado.

`plot()` puede tomar argumentos que se pasan a matplotlib para diseñar la trama de diferentes maneras.

```python
# Comente linea por linea para ver las diferencias.

df.plot(style='o') # trazado como puntos
df.plot(style= 'g--') # trazado como linea verde punteada
df.plot(style= '--o', markeredgecolor='white') # trazado como puntos con borde blanco
```

#### Trazar en un eje matplotlib existente.

Por defecto, `plot()` crea una nueva figura cada vez que se llama. Es posible trazar en un eje existente pasando el parámetro `ax`.

```python
df1 = pd.DataFrame(
    {
        'x': [10,8,10,7,7,10,9,9],
        'y': [6,4,5,5,7,10,9,9]
    }
)

df2 = pd.DataFrame(
    {
        'x': [10,8,10,7,7,10,9,9],
        'y': [9,9,10,7,5,5,4,6]
    }
)

plt.figure() # crea una nueva figura
ax = plt.subplot(121) # crea un subplot en el lado izquiedo
df1.plot(ax=ax) # grafica el dataframe en el subplot creado
ax = plt.subplot(122) # crea un subplot en el lado derecho
df2.plot(ax=ax) # grafica el dataframe en el subplot creado
plt.show() # muestra la nueva figura
```

```python
g = a - b # Resta de matrices
np.subtract(a, b)
add = a + b # Suma de matrices
np.add(a, b)
divide = a / b # Division de matrices
np.divide(a, b)
multiply = a * b # Multiplicacion de matrices
np.multiply(a, b)
power = a ** b # Potencia de matrices
np.power(a, b)
np.exp(a)
np.sqrt(a)
np.sin(a)
np.cos(a)
np.log(a)
np.abs(a)
e.dot(f) # Producto interno de dos matrices
np.dot(e, f) # Producto interno de dos matrices
np.cross(a, b) # Producto cruzado de dos vectores
np.transpose(a) # Transpuesta de una matriz
np.linalg.det(a) # Determinante de una matriz
np.linalg.inv(a) # Inversa de una matriz
np.linalg.eig(a) # Valores y vectores propios de una matriz
np.linalg.svd(a) # SVD de una matriz
np.linalg.qr(a) # QR factorization de una matriz
np.linalg.solve(a, b) # Solucion de un sistema de ecuaciones
```

```python
a == b # Comparacion de matrices igualdad
np.equal(a, b)
a != b # Comparacion de matrices desigualdad
np.not_equal(a, b)
a < b # Comparacion de matrices menor que
np.less(a, b)
a > b # Comparacion de matrices mayor que
np.greater(a, b)
a <= b # Comparacion de matrices menor o igual que
np.less_equal(a, b)
a >= b # Comparacion de matrices mayor o igual que
np.greater_equal(a, b)
```

## gradiente
`numpy.gradiente(f, *varargs, eje=None, edge_orden=1)[fuente]`

Devuelve el gradiente de una matriz N-dimensional.

El gradiente se calcula utilizando diferencias centrales precisas de segundo orden en los puntos interiores y en un lado preciso de primer o segundo orden (hacia adelante o hacia atr√°s) diferencias en los l√≠mites. El devuelto gradiente por lo tanto, tiene la misma forma que la matriz de entrada.

## Par√°metros:

### f array_like

Una matriz N-dimensional que contiene muestras de una funci√≥n escalar.

### varargs: *lista de escalar o array, opcional*

Espaciado entre valores f. Espacio unitario predeterminado para todas las dimensiones. El espaciado se puede especificar usando:

1. escalar √∫nico para especificar una distancia de muestra para todas las dimensiones.

2. N escalares para especificar una distancia de muestra constante para cada dimensi√≥n. i.e. dx, dy, dz, ‚Ä¶

3. N matrices para especificar las coordenadas de los valores a lo largo de cada dimensi√≥n de F. La longitud de la matriz debe coincidir con el tama√±o de la dimensi√≥n correspondiente

4. Cualquier combinaci√≥n de escalares / arrays N con el significado de 2. y 3.

Si eje se da, el n√∫mero de varargs debe ser igual al n√∫mero de ejes. Predeterminado: 1.

### edge_order{ 1, 2 }, opcional
Gradiente se calcula utilizando diferencias precisas de orden N-√©simo en los l√≠mites. Predeterminado: 1.

> Nuevo en la versi√≥n 1.9.1.

### axis: Ninguno o int o tupla de ints, opcional
Gradiente se calcula solo a lo largo del eje o ejes dados El eje predeterminado ( = Ninguno ) es calcular el gradiente para todos los ejes de la matriz de entrada. El eje puede ser negativo, en cuyo caso cuenta desde el √∫ltimo al primer eje.

> Nuevo en la versi√≥n 1.11.0.

## Return:

### gradiente : ndarray o lista de ndarray

Una lista de ndarrays (o un solo ndarray si solo hay una dimensi√≥n) correspondiente a las derivadas de f con respecto a cada dimensi√≥n. Cada derivado tiene la misma forma que f.



## Notas

Asumiendo eso $f \epsilon C^3$
(es decir., $f$ tiene al menos 3 continuos derivados) y let $h_*$ sea un tama√±o de paso no homog√©neo, nosotros minimiza el error ‚Äúconsistence‚Äù $n_i$
 entre lo verdadero gradiente y su estimaci√≥n de una combinaci√≥n lineal de los puntos de cuadr√≠cula vecinos:

Sustituyendo $f(x_i + h_d)$ y $f(x_1 - h_s)$ con su expansi√≥n de la serie Taylor, esto se traduce en resoluci√≥n el siguiente sistema lineal:

 $${\alpha + \beta + \gamma = 0}{\beta h_d - \gamma h_s = 1}{\beta h_d^2 + \gamma h_s^2 = 0}$$

La aproximaci√≥n resultante de $f_i^1$ es lo siguiente: 

$$f_i^1 = \frac{h_s^2(x_i + h_d) + (h_d^2 - h_s^2) f(x_i) - h_d^2f(x_i - h_s)}{h_sh_d(h_d + h_s)} + O \frac{h_dh_s^2 + h_sh_d^2}{h_d + h_s}$$

Vale la pena se√±alar que si $h_s = h_d$
 (es decir, los datos est√°n espaciados uniformemente) encontramos la aproximaci√≥n est√°ndar de segundo orden:

 $$f_i^1 = \frac{f(x_i + 1) - f(x_i - 1)}{2h} + O(h^2)$$

Con un procedimiento similar, las aproximaciones hacia delante/hacia atr√°s utilizadas para se pueden derivar l√≠mites.



### Referencias

1. Quarteroni A., Sacco R., Saleri F. (2007) Matem√°ticas Num√©ricas (Textos en Matem√°ticas Aplicadas). Nueva York: Springer.

2. Durran D. R. (1999) M√©todos Num√©ricos para Ecuaciones de Ondas en Geophysical Fluid Dynamics. Nueva York: Springer.

3. B Fornberg. (1988) Generaci√≥n de F√≥rmulas de Diferencia Finita en Rejillas Arbitrariamente Espaciadas, Matem√°ticas de la computaci√≥n 51, no. 184:699-706. PDF.

### Ejemplos


```python
import numpy as np

f = np.array([1, 2, 4, 7, 11, 16], dtype=float)
np.gradient(f)
# array([1. , 1.5, 2.5, 3.5, 4.5, 5. ])
```

```python
np.gradient(f, 2)
# array([0.5 ,  0.75,  1.25,  1.75,  2.25,  2.5 ])
```

El espaciado tambi√©n se puede especificar con una matriz que representa las coordenadas de los valores F a lo largo de las dimensiones. Por ejemplo, un espaciado uniforme:


```python

x = np.arange(f.size)
np.gradient(f, x)
# array([1. ,  1.5,  2.5,  3.5,  4.5,  5. ])
```

O uno no uniforme:


```python

x = np.array([0., 1., 1.5, 3.5, 4., 6.], dtype=float)
np.gradient(f, x)
# array([1. ,  3. ,  3.5,  6.7,  6.9,  2.5])
```

Para matrices bidimensionales, el retorno ser√° de dos matrices ordenadas por eje. En este ejemplo, la primera matriz representa el gradiente en filas y la segunda en direcci√≥n de columnas:


```python

np.gradient(np.array([[1, 2, 6], [3, 4, 5]], dtype=float))
# [array([[ 2.,  2., -1.],[ 2.,  2., -1.]]), array([[1. , 2.5, 4. ], [1. , 1. , 1. ]])]
```

En este ejemplo tambi√©n se especifica el espaciado: uniforme para axis=0 y no uniforme para axis=1


```python

dx = 2.
y = [1., 1.5, 3.5]
np.gradient(np.array([[1, 2, 6], [3, 4, 5]], dtype=float), dx, y)
# [array([[ 1. ,  1. , -0.5],[ 1. ,  1. , -0.5]]), array([[2. , 2. , 2. ], [2. , 1.7, 0.5]])]
```

Es posible especificar c√≥mo se tratan los l√≠mites usando edge_orden


```python

x = np.array([0, 1, 2, 3, 4])
f = x**2
np.gradient(f, edge_order=1)
# array([1.,  2.,  4.,  6.,  7.])
```

```python

np.gradient(f, edge_order=2)
# array([0., 2., 4., 6., 8.])
```

El eje la palabra clave se puede utilizar para especificar un subconjunto de ejes de los cuales el gradiente se calcula


```python

np.gradient(np.array([[1, 2, 6], [3, 4, 5]], dtype=float), axis=0)
# array([[ 2.,  2., -1.],       [ 2.,  2., -1.]])
```

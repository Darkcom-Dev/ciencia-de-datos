```python
import torch
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dataset/auto+mpg/auto-mpg.data', comment='\t', names=None, sep= " ",  skipinitialspace=True)
```

```python
df
```

```python
df.shape
```

```python
df.columns= ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration','model year' ,'origin']
```

```python
df
```

```python
df =df.replace('?', np.NaN)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df =df.reset_index(drop=True)
```

```python
df.info()
```

```python
intervalos = [float("-inf"), 76, 79, float("inf")]

# Crear una nueva columna con los nÃºmeros de intervalos
df['bins'] = pd.cut(df['model year'], bins=intervalos, labels=False, right=True)

# Mostrar el DataFrame con la nueva columna
df['bins']
```

```python
one_hot_encoded = pd.get_dummies(df['bins'], dtype=int)
one_hot_encoded
```

```python
df = pd.concat([df, one_hot_encoded], axis=1)
df
```

```python
df = df.drop(columns=['model year', 'bins', 2]) # 'bins'  y 'model year' sobra y '2' es combinacion lineal de 0 y 1
df.columns= ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration','origin', 'grupo 0', 'grupo 1']
df
```

```python
X = pd.concat([df.iloc[:, :6], df.iloc[:, 7:]], axis=1)
X
```
